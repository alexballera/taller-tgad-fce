### 1.3)

Definir una función llamada `clasificar_riesgo` que reciba `deuda` e `ingreso` y retorne la categoría correspondiente:

| Condición | Categoría |
|---|---|
| ratio < 0.25 | `SALUDABLE` |
| 0.25 ≤ ratio ≤ 0.45 | `MODERADO` |
| ratio > 0.45 | `RIESGOSO` |


Luego recorriendo `solicitantes` llamar a la función para cada uno, imprimí nombre y categoría, y agregar la categoría al diccionario del solicitante bajo la clave `"categoria"`.

````python
# 1.3 — Función de clasificación de riesgo

def clasificar_riesgo(deuda, ingreso):
    """
    Clasifica el riesgo crediticio basado en la relación deuda/ingreso.
    """
    ratio = deuda / ingreso
    if ratio < 0.25:
        return "SALUDABLE"
    elif ratio <= 0.45:
        return "MODERADO"
    else:
        return "RIESGOSO"

# Recorrer solicitantes y agregar categoría
for solicitante in solicitantes:
    categoria = clasificar_riesgo(solicitante["deuda_actual"], solicitante["ingreso_anual"])
    solicitante["categoria"] = categoria
    print(f"{solicitante['nombre']} — {categoria}")
````

Definir una función llamada `reporte_solicitante` que reciba:
- `nombre` (str)
- `deuda` (float)
- `ingreso` (float)
- `detallado` (bool) — **argumento optativo**, valor por defecto `False`

Comportamiento:
- `detallado=False` → imprime: `"Ana López — MODERADO"`
- `detallado=True`  → imprime: `"Ana López — MODERADO  (ratio: 0.40  |  ingreso: 45000)"`

La función debe usar internamente `clasificar_riesgo` del Ejercicio 1.

````python
# 2.1 — Función con argumento optativo
def reporte_solicitante(nombre, deuda, ingreso, detallado=False):
    """
    Imprime un reporte de riesgo para un solicitante.
    Usa la función clasificar_riesgo definida en el Ejercicio 1.
    """
    categoria = clasificar_riesgo(deuda, ingreso)
    
    if detallado:
        ratio = deuda / ingreso
        print(f"{nombre} — {categoria}  (ratio: {ratio:.2f}  |  ingreso: {ingreso})")
    else:
        print(f"{nombre} — {categoria}")
````

````python
# 2.3 — Clase Solicitante con métodos de cálculo y clasificación

class Solicitante:
    def __init__(self, nombre, ingreso_anual, deuda_actual):
        self.nombre        = nombre
        self.ingreso_anual = ingreso_anual
        self.deuda_actual  = deuda_actual

    def calcular_ratio(self):
        """Calcula el ratio deuda/ingreso."""
        return self.deuda_actual / self.ingreso_anual

    def clasificar(self):
        """Clasifica el riesgo crediticio usando la lógica de clasificar_riesgo."""
        return clasificar_riesgo(self.deuda_actual, self.ingreso_anual)

# Instanciar 2 objetos y mostrar nombre, ratio y clasificación
sol1 = Solicitante("Juan García", 60000, 15000)
sol2 = Solicitante("María López", 50000, 30000)

for sol in [sol1, sol2]:
    print(f"{sol.nombre} — Ratio: {sol.calcular_ratio():.2f} — {sol.clasificar()}")
````

### 4.2)

Obtener la cantidad de solicitudes por cada combinación de `product_type` y `loan_intent`.

Responder con un comentario: ¿Cuál es la combinación más frecuente? ¿Cuál la menos frecuente?

````python
# 4.2 — Cantidad de solicitudes por combinación de product_type y loan_intent

combinaciones = (
    DATA.groupby(["product_type", "loan_intent"])
    .size()
    .reset_index(name="cantidad_solicitudes")
    .sort_values("cantidad_solicitudes", ascending=False)
)

display(combinaciones)

# Identificar combinación(es) más y menos frecuente(s)
max_cant = combinaciones["cantidad_solicitudes"].max()
min_cant = combinaciones["cantidad_solicitudes"].min()

mas_frecuentes = combinaciones[combinaciones["cantidad_solicitudes"] == max_cant]
menos_frecuentes = combinaciones[combinaciones["cantidad_solicitudes"] == min_cant]

print("\nCombinación(es) más frecuente(s):")
display(mas_frecuentes)

print("Combinación(es) menos frecuente(s):")
display(menos_frecuentes)

# Comentario:
# La(s) combinación(es) mostrada(s) en "más frecuente(s)" es/son la(s) de mayor cantidad de solicitudes,
# y la(s) de "menos frecuente(s)" es/son la(s) de menor cantidad en el dataset.
````
