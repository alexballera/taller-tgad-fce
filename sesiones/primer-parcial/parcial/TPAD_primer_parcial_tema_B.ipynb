### 1.2)

Crear una función llamada `investigar` que reciba un DataFrame y muestre:
- Cantidad de filas y columnas
- Nombres de columnas
- Tipo de dato de cada columna
- Porcentaje de columnas por tipo

Aplicar sobre `DATA`.

````python
# 1.2 — Función para investigar la estructura del DataFrame
def investigar(df):
    filas, columnas = df.shape
    print(f"Cantidad de filas: {filas}")
    print(f"Cantidad de columnas: {columnas}")

    print("\nNombres de columnas:")
    print(list(df.columns))

    print("\nTipo de dato de cada columna:")
    display(df.dtypes.rename("tipo_dato"))

    print("\nPorcentaje de columnas por tipo:")
    porcentaje_tipos = (df.dtypes.value_counts(normalize=True) * 100).round(2)
    display(porcentaje_tipos.rename("porcentaje"))

# Aplicar sobre DATA
investigar(DATA)
````

### 1.3)

Detectar si existen valores faltantes (`NaN`) en el DataFrame. Mostrar el porcentaje de valores faltantes por cada columna.

````python
# 1.3 — Detección de valores faltantes (NaN)
porcentaje_faltantes = (DATA.isna().mean() * 100).round(2)

if porcentaje_faltantes.gt(0).any(): # evalúo si alguna columna tiene más de 0% de valores faltantes
    print("Sí, existen valores faltantes en el DataFrame.")
else:
    print("No existen valores faltantes en el DataFrame.")

print("\nPorcentaje de valores faltantes por columna:")
display(porcentaje_faltantes.rename("porcentaje_faltantes"))
````

### 2.2)

Obtener la cantidad de solicitudes por cada combinación de `product_type` y `loan_intent`.

Responder con un comentario: ¿Cuál es la combinación más frecuente? ¿Cuál la menos frecuente?

````python
# 2.2 — Cantidad de solicitudes por combinación de product_type y loan_intent
conteo_combinaciones = (
    DATA.groupby(["product_type", "loan_intent"])
    .size()
    .reset_index(name="cantidad_solicitudes")
    .sort_values("cantidad_solicitudes", ascending=False)
    .reset_index(drop=True)
)

display(conteo_combinaciones)

# Combinación más y menos frecuente
mas_frecuente = conteo_combinaciones.iloc[0]
menos_frecuente = conteo_combinaciones.iloc[-1]

print(
    f"Más frecuente: ({mas_frecuente['product_type']}, {mas_frecuente['loan_intent']}) "
    f"con {mas_frecuente['cantidad_solicitudes']} solicitudes."
)
print(
    f"Menos frecuente: ({menos_frecuente['product_type']}, {menos_frecuente['loan_intent']}) "
    f"con {menos_frecuente['cantidad_solicitudes']} solicitudes."
)
````
