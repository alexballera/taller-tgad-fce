##**2. STRINGS**


Estos son una cadena de caracteres. Para crear un string se debe encerrar la cadena entre comillas dobles. También es posible entre comillas simples, pero en el caso del idioma inglés por ejemplo, nos dará error si en el texto hay una comilla simple. Frente a esto es mejor utilizar siempre comillas dobles.

```python
#definimos un string
"Taller de programación"
```

```python
# o podemos imprimirlo
print("Taller de programación")
```

```python
# o podemos asignarlo a una variable y luego llamarla o imprimirla
texto="Taller de programación"
texto
```

```python
print(texto)
```

Funciones más frecuentemente utilizadas con strings

```python
# cantidad de elementos de la cadena (incluye los espacios en blanco)
len(texto)
```

```python
#pasar a mayúsculas
texto.upper()
```

```python
# pasar a minúscula
texto.lower()
```

```python
# separar la cadena por cada espacio en blanco
texto.split()
```

```python
# separar la cadena por alguna letra
texto.split("a")
```

Al ser una cadena de texto, contiene varios elementos (cada letra será un elemento, pero el espacio en blanco también). Entonces si queremos obtener un elemento en particular de la cadena utilizamos el indice.
En Python, la posición de los elemento comienza en cero.

```python
texto="Taller de programación"
```

```python
texto
```

```python
# la primer letra
texto[0]
```

```python
# algún elemento de la cadena
texto[2]
```

```python
# el útlimo elemento de la cadena
texto[-1]
```

```python
# si quiero obtener todo lo anterior a una posición determinada
texto[:18]
```

```python
# o todo lo posterior
texto[18:]
```

```python
# si quiero obtener desde una posición a otra
texto[0:6]
```

Algo común de realizar es la concatenación

```python
texto + ' 2021'
# también podríamos asignarlo a una variable y luego llamarla
```

```python
# si quisiereamos generar una repetición:
x="Z"
x*10
```

```python
x=texto + ' 2021'
x
```

```python
x=texto + ' 2021'
x*10
```

##**3.BOOLEANS**

Este tipo solo puede devolver verdadero o falso. Resultan útiles para cuando se quiere establecer una condición en un proceso lógico (IF, WHILE que veremos en la próxima clase)




<table class="table table-bordered">
<tr>
<th style="width:10%">Operador</th><th style="width:45%">Descripción</th><th>Example</th>
</tr>
<tr>
<td>==</td>
<td>Si los valores son exactamente igual, la condición será verdadera.</td>
<td> (a == b) is not true.</td>
</tr>
<tr>
<td>!=</td>
<td>Si los valores son distintos , la condición será verdadera.</td>
<td> (a != b) is true.</td>
</tr>
<tr>
<td>&gt;</td>
<td>Si el valor de la izquierda es estrictamente mayor que el de la derecha, la condición devolverá verdadero.</td>
<td> (a &gt; b) is not true.</td>
</tr>
<tr>
<td>&lt;</td>
<td>Si el valor de la izquierda es estrictamente menor que el de la derecha, la condición devolverá verdadero.</td>
<td> (a &lt; b) is true.</td>
</tr>
<tr>
<td>&gt;=</td>
<td>Si el valor de la izquierda es mayor o igual que el de la derecha, la condición devolverá verdadero.</td>
<td> (a &gt;= b) is not true. </td>
</tr>
<tr>
<td>&lt;=</td>
<td>Si el valor de la izquierda es menosr o igual que el de la derecha, la condición devolverá verdadero.</td>
<td> (a &lt;= b) is true. </td>
</tr>
</table>

```python
1>2 #preguntamos si uno es mayor que dos
```

```python
2>=1 #preguntamos si dos es mayor o igual que uno
```

```python
2==1 #preguntamos si dos es igual a uno
```
