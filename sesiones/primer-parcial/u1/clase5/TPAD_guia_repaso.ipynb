### Ejercicio 1.1 — Ficha de Producto

Una empresa de distribución necesita generar la ficha técnica de un artículo.

**Consignas:**
1. Calcular el `precio_final` aplicando IVA (21%) si `aplica_iva` es `True`.
2. Calcular el `valor_stock` (precio_final × stock).
3. Imprimir la ficha con el formato del ejemplo.

**Salida esperada:**
```
Código  : 10042
Nombre  : Resma A4 500 hojas
Precio  : $ 1.452,30
Stock   : 8 unidades
Val.inv.: $ 11.618,42
```
> **Pista:** para convertir el punto decimal al formato argentino, usá `.replace()`.

````python
codigo      = 10042
nombre      = "Resma A4 500 hojas"
precio_neto = 1200.25
aplica_iva  = True
stock       = 8
TASA_IVA    = 0.21

# 1. Precio final
if aplica_iva:
    precio_final = precio_neto * (1 + TASA_IVA)
else:
    precio_final = precio_neto


# 2. Valor del stock
valor_stock = precio_final * stock

# 3. Imprimir la ficha
print(f"Código: {codigo}")
print(f"Nombre: {nombre}")
print(f"Precio neto: {precio_neto:.2f}".replace('.', ','))
print(f"Aplica IVA: {aplica_iva}")
print(f"Precio final: {precio_final:.2f}".replace('.', ','))
print(f"Stock: {stock}")
print(f"Valor del stock: {valor_stock:.2f}".replace('.', ','))
````

### Ejercicio 1.2 — Ticket de Caja

En un sistema de punto de venta los datos llegan como **strings**. Tu tarea es procesarlos.

**Consignas:**
1. Convertir `precio_str` a float (el formato `"4.599,90"` usa punto de miles y coma decimal).
2. Si el cliente es VIP (`"si"` en cualquier capitalización), sumale 5 puntos al descuento.
3. Calcular `subtotal`, `monto_descuento` y `total_con_iva` (21%). Redondear a 2 decimales.
4. Imprimir los valores con formato.

> **Pista:** `"4.599,90".replace('.', '').replace(',', '.')` devuelve `"4599.90"`.

````python
cantidad_str  = "3"
precio_str    = "4.599,90"   # punto de miles, coma decimal
descuento_str = "15"          # porcentaje
cliente_vip   = "Si"

# 1. Convertir precio_str a float
precio_unitario = float(precio_str.replace('.', '').replace(',', '.'))

# 2. Descuento final (+5 pts si es VIP)
descuento_pct = float(descuento_str) + 5 if cliente_vip == "Si" else float(descuento_str)

# 3. Cálculos (redondeá con round())
cantidad        = int(cantidad_str)
subtotal        = round(cantidad * precio_unitario, 2)
monto_descuento = round(subtotal * descuento_pct / 100, 2)
base_iva        = round(subtotal - monto_descuento, 2)
monto_iva       = round(base_iva * 0.21, 2)
total_con_iva   = round(base_iva + monto_iva, 2)

# 4. Imprimir
# TU CÓDIGO AQUÍ
print(f"Cantidad: {cantidad}")
print(f"Precio unitario: {precio_unitario}")
print(f"Subtotal: {subtotal}")
print(f"Monto de descuento: {monto_descuento}")
print(f"Base IVA: {base_iva}")
print(f"Monto IVA: {monto_iva}")
print(f"Total con IVA: {total_con_iva}")
````

### Ejercicio 3.1 — Clasificación de Clientes

Una empresa clasifica a sus clientes según su facturación anual y les asigna un descuento.

| Categoría | Facturación anual | Descuento |
|-----------|-------------------|-----------|
| Platinum  | ≥ $5.000.000      | 20%       |
| Gold      | ≥ $1.000.000      | 12%       |
| Silver    | ≥ $300.000        | 6%        |
| Regular   | < $300.000        | 0%        |

**Consignas:**
1. Escribir la función `clasificar(facturacion)` que retorne `(categoria, descuento_pct)`.
2. Recorrer la lista de clientes, clasificar a cada uno y calcular su precio final para una compra dada.
3. Imprimirs un resumen con todos los datos.
4. **Extra:** mostrá cuántos clientes subirían de categoría si incrementaran su facturación un 20%.

````python
clientes = [
    ("Distribuidora Sur SA",   6_200_000, 150_000),
    ("Comercio El Rápido",       850_000,  28_000),
    ("Consultora Números",     1_450_000,  95_000),
    ("Kiosco Don Pedro",         180_000,   4_500),
    ("Supermercado Central",   3_900_000, 210_000),
    ("Ferretería del Norte",     320_000,  18_700),
]
# (nombre, facturación_anual, compra_actual)

def clasificar(facturacion):
    """
    Retorna (categoria, descuento_pct) según la facturación anual.
    """
    if facturacion >= 5_000_000:
        return "Platinum", 20
    elif facturacion >= 1_000_000:
        return "Gold", 12
    elif facturacion >= 300_000:
        return "Silver", 6
    else:
        return "Regular", 0

# Procesar cada cliente
print(f"{'Cliente':<26} {'Cat.':<10} {'Dto':>4}  {'Compra':>10}  {'Total':>10}")
print("-" * 70)
for nombre, facturacion, compra in clientes:
    categoria, dto = clasificar(facturacion)
    total = round(compra * (1 - dto / 100), 2)
    print(f"{nombre:<26} {categoria:<10} {dto:>4}%  ${compra:>10,.2f}  ${total:>10,.2f}".replace('.', ','))

# Extra: ¿cuántos subirían de categoría con +20%?
incremento = 0.20
print("\nClientes que subirían de categoría con +20% de facturación:")
for nombre, facturacion, compra in clientes:
    categoria_actual, _ = clasificar(facturacion)
    categoria_nueva, _ = clasificar(facturacion * (1 + incremento))
    if categoria_nueva != categoria_actual:
        print(f"- {nombre} subiría de {categoria_actual} a {categoria_nueva}")
````

### Ejercicio 3.2 — Liquidación de Sueldos

El área de RRHH necesita calcular el sueldo neto de cada empleado según su categoría y horas trabajadas.

**Reglas de liquidación:**

| Categoría | Valor hora | Horas extra (más de 160 h/mes) |
|-----------|-----------|--------------------------------|
| Junior    | $ 1.800   | +50% sobre el valor hora       |
| Semi      | $ 2.800   | +50% sobre el valor hora       |
| Senior    | $ 4.200   | +75% sobre el valor hora       |

- Las primeras 160 horas se pagan al valor normal.
- Las horas adicionales se pagan al valor de hora multiplicado por `(1 + extra_pct)`.
- Se descuenta **17% de aportes** sobre el bruto total.

**Consignas:**
1. Implementar `calcular_sueldo(categoria, horas)` que retorne `(bruto, aportes, neto)`.
2. Procesar todos los empleados e imprimí el resumen.
3. Mostrar al empleado con mayor sueldo neto y el costo total para la empresa.

````python
empleados = [
    ("Gómez, Laura",   "Senior", 172),
    ("Pérez, Carlos",  "Junior", 160),
    ("Díaz, Sofía",    "Semi",   185),
    ("Torres, Martín", "Senior", 160),
    ("López, Ana",     "Junior", 168),
    ("Ruiz, Pablo",    "Semi",   155),
]
# (nombre, categoría, horas_trabajadas)

VALOR_HORA = {"Junior": 1_800, "Semi": 2_800, "Senior": 4_200}
EXTRA_PCT  = {"Junior": 0.50,  "Semi": 0.50,  "Senior": 0.75}
HORAS_BASE = 160
APORTES    = 0.17

def calcular_sueldo(categoria, horas):
    """
    Retorna (bruto, aportes, neto).
    Las primeras HORAS_BASE se pagan al valor normal;
    el excedente se paga al valor * (1 + EXTRA_PCT).
    """
    valor_hora = VALOR_HORA[categoria]
    extra_pct = EXTRA_PCT[categoria]

    if horas <= HORAS_BASE:
        bruto = horas * valor_hora
    else:
        bruto = HORAS_BASE * valor_hora + (horas - HORAS_BASE) * valor_hora * (1 + extra_pct)

    aportes = round(bruto * APORTES, 2)
    neto = round(bruto - aportes, 2)
    return bruto, aportes, neto

print(f"{'Empleado':<18} {'Cat.':<8} {'Hs':>4}  {'Bruto':>11}  {'Aportes':>11}  {'Neto':>11}")
print("-" * 72)

# recorrés los empleados, llamar calcular_sueldo y mostrás
for nombre, categoria, horas in empleados:
    bruto, aportes, neto = calcular_sueldo(categoria, horas)
    print(f"{nombre:<18} {categoria:<8} {horas:>4}  ${bruto:>11,.2f}  ${aportes:>11,.2f}  ${neto:>11,.2f}".replace('.', ','))

# Empleado con mayor neto y costo total
empleado_mayor_neto = None
mayor_neto = float('-inf')
costo_total = 0

for nombre, categoria, horas in empleados:
    bruto, aportes, neto = calcular_sueldo(categoria, horas)
    costo_total += bruto
    if neto > mayor_neto:
        mayor_neto = neto
        empleado_mayor_neto = nombre
print("-" * 72)
print(f"Empleado con mayor neto: {empleado_mayor_neto}")
print(f"Costo total de sueldos: ${costo_total:>11,.2f}".replace('.', ','))
````

### Ejercicio 3.3 — Sistema de Facturación

Implementar un módulo de facturación que combine clasificación de clientes, validación y generación de comprobantes.

**Reglas del sistema:**
- El descuento de cada cliente se determina con `clasificar()` del ejercicio 3.1.
- Una factura solo puede emitirse si el subtotal con descuento supera `$1.000`.
- El IVA (21%) se calcula sobre el subtotal ya descontado.

**Funciones a implementar:**

| Función | Retorna |
|---------|--------|
| `calcular_factura(cliente, items)` | `(subtotal, monto_dto, base_iva, monto_iva, total)` |
| `es_valida(base_iva)` | `bool` (True si base_iva ≥ $1.000) |
| `imprimir_factura(cliente, items)` | imprime el comprobante o un mensaje de error |

**Consignas:**
1. Implementar las tres funciones. `calcular_factura` debe usar `clasificar()` internamente.
2. Probar `imprimir_factura` con los tres casos definidos.
3. Mostrar el total facturado sumando solo los comprobantes válidos.

> **Clave del ejercicio:** las funciones se encadenan — `imprimir_factura` llama a `calcular_factura` que llama a `clasificar`. Así es como se estructura código real.

````python
# Reutilizar clasificar() del ejercicio 3.1 — asegurate de haberlo ejecutado.

casos = [
    (
        {"nombre": "Distribuidora Sur SA", "facturacion_anual": 6_200_000},
        [("Resma A4 x500", 20, 1_450), ("Lapicera BIC x10", 10, 550)]
    ),
    (
        {"nombre": "Kiosco Don Pedro", "facturacion_anual": 180_000},
        [("Corrector líquido", 2, 280)]   # total < $1.000 → inválida
    ),
    (
        {"nombre": "Consultora Números", "facturacion_anual": 1_450_000},
        [("Monitor 24''", 3, 58_400), ("Mouse USB", 3, 3_200)]
    ),
]
# cliente = {"nombre": str, "facturacion_anual": int}
# items   = [(descripcion, cantidad, precio_unitario), ...]

def calcular_factura(cliente, items):
    """
    Retorna (subtotal, monto_dto, base_iva, monto_iva, total).
    Usa clasificar() para obtener el descuento del cliente.
    """
    subtotal = sum(cantidad * precio_unitario for _, cantidad, precio_unitario in items)
    _, descuento_pct = clasificar(cliente["facturacion_anual"])
    monto_dto = round(subtotal * descuento_pct / 100, 2)
    base_iva = round(subtotal - monto_dto, 2)
    monto_iva = round(base_iva * 0.21, 2)
    total = round(base_iva + monto_iva, 2)

    return subtotal, monto_dto, base_iva, monto_iva, total


def es_valida(base_iva):
    """Retorna True si base_iva es mayor o igual a $1.000."""
    return base_iva >= 1000


def imprimir_factura(cliente, items):
    """Imprime el comprobante completo, o un mensaje de error si no es válida."""
    subtotal, monto_dto, base_iva, monto_iva, total = calcular_factura(cliente, items)

    if not es_valida(base_iva):
        print(f"Factura inválida para {cliente['nombre']}: base IVA menor a $1.000")
        return

    print(f"Factura para {cliente['nombre']}")
    print("-" * 40)
    for descripcion, cantidad, precio_unitario in items:
        print(f"{descripcion:<20} {cantidad:>3} x ${precio_unitario:>10,.2f}".replace('.', ','))
    print("-" * 40)
    print(f"{'Subtotal:':<30} ${subtotal:>10,.2f}".replace('.', ','))
    print(f"{'Descuento:':<30} -${monto_dto:>10,.2f}".replace('.', ','))
    print(f"{'Base IVA:':<30} ${base_iva:>10,.2f}".replace('.', ','))
    print(f"{'IVA (21%):':<30} ${monto_iva:>10,.2f}".replace('.', ','))
    print(f"{'Total a pagar:':<30} ${total:>10,.2f}".replace('.', ','))


# Probar con los tres casos
for cliente, items in casos:
    imprimir_factura(cliente, items)
    print()

# Total facturado (solo comprobantes válidos)
total_facturado = 0
for cliente, items in casos:
    _, _, base_iva, _, total = calcular_factura(cliente, items)
    if es_valida(base_iva):
        total_facturado += total

print(f"Total facturado: ${total_facturado:,.2f}".replace('.', ','))
````
