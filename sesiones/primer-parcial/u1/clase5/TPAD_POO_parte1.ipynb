### **Programación Orientada a Objetos (POO) — Parte 1**

En esta notebook vamos a ver una forma diferente de organizar nuestros programas: la **Programación Orientada a Objetos**.

Hasta ahora escribimos código de forma **procedural**: una serie de instrucciones que se ejecutan una tras otra, usando variables sueltas y funciones. La POO nos da una nueva herramienta: podemos **agrupar datos y comportamiento** en una sola entidad llamada **objeto**.

En esta primera parte veremos:
- ¿Qué son las clases y los objetos?
- Cómo definir una clase
- El constructor `__init__` y el parámetro `self`
- Atributos: las variables del objeto
- Métodos: las funciones del objeto
- El método especial `__str__`
- Type hints y docstrings en clases

---
### **¿Qué son las clases y los objetos?**

Pensemos en el mundo real. Un **alumno** tiene:
- **Datos** (atributos): nombre, legajo, notas
- **Acciones** (métodos): calcular promedio, aprobar o desaprobar

En el enfoque **procedural**, modelaríamos esto con variables sueltas y funciones separadas:

```python
nombre = "Ana"
notas  = [8.0, 7.5, 9.0]

def calcular_promedio(notas):
    ...
```

En el enfoque **orientado a objetos**, todo eso vive junto en una sola entidad:

```python
alumno = Alumno(nombre="Ana", notas=[8.0, 7.5, 9.0])
alumno.calcular_promedio()
```

| Concepto | Analogía | En Python |
|---|---|---|
| **Clase** | El molde / plano de una casa | `class Alumno:` |
| **Objeto** | Una casa construida con ese molde | `alumno = Alumno(...)` |
| **Atributo** | Las características del objeto | `alumno.nombre` |
| **Método** | Las acciones que puede hacer | `alumno.calcular_promedio()` |

> Con una misma **clase** podemos crear infinitos **objetos**, cada uno con sus propios datos.

````python
# Ejemplo conceptual: con el mismo molde creamos distintos objetos
# (lo veremos en detalle en las próximas secciones)

# alumno1 = Alumno(nombre="Ana",   legajo=1001)
# alumno2 = Alumno(nombre="Bruno", legajo=1002)
# alumno3 = Alumno(nombre="Carla", legajo=1003)

# Los tres son objetos del mismo tipo (Alumno)
# pero cada uno tiene sus propios datos
print("Una clase = un molde. Muchos objetos = muchas instancias.")
````

### **Definir una clase**

Para definir una clase usamos la palabra clave `class` seguida del nombre y dos puntos.

```python
class NombreDeLaClase:
    # atributos y métodos
    ...
```

**Convención de nombres:**
- Las clases usan **PascalCase**: primera letra de cada palabra en mayúscula
- `Alumno`, `ProductoVenta`, `CuentaBancaria` ✓
- `alumno`, `producto_venta`, `cuenta_bancaria` ✗ (eso es para variables y funciones)

````python
# La clase más simple posible
class Alumno:
    pass   # 'pass' indica que el cuerpo está vacío por ahora

# Creamos un objeto (instancia) de la clase
alumno1 = Alumno()

print(alumno1)
print(type(alumno1))
````

### **El constructor `__init__` y `self`**

Cuando creamos un objeto, Python llama automáticamente a un método especial llamado `__init__` (doble guión bajo antes y después). Este método se llama **constructor** y sirve para **inicializar los atributos** del objeto.

```python
class MiClase:
    def __init__(self, param1, param2):
        self.atributo1 = param1
        self.atributo2 = param2
```

#### ¿Qué es `self`?

`self` es una referencia al **propio objeto** que se está creando o usando. Es el primer parámetro de todo método, pero **no lo pasamos nosotros** al llamar: Python lo pasa automáticamente.

- `self.nombre = nombre` → estamos guardando el valor en **este objeto específico**
- Sin `self`, la variable sería local a `__init__` y desaparecería al terminar

> ⚠️ `self` no es una palabra reservada, pero es la convención universal. Siempre usá `self`.

````python
class Alumno:
    def __init__(self, nombre: str, legajo: int):
        self.nombre: str = nombre    # atributo del objeto
        self.legajo: int = legajo    # atributo del objeto

# Al crear el objeto, pasamos los argumentos de __init__ (sin self)
alumno1 = Alumno("Ana", 1001)
alumno2 = Alumno("Bruno", 1002)

# Cada objeto tiene sus propios datos
print(alumno1.nombre, alumno1.legajo)
print(alumno2.nombre, alumno2.legajo)
````

````python
# Acceder y modificar atributos
alumno1 = Alumno("Ana", 1001)

print("Nombre original:", alumno1.nombre)

# Podemos modificar un atributo directamente
alumno1.nombre = "Ana García"
print("Nombre actualizado:", alumno1.nombre)
````

#### `self` en detalle: por qué es necesario

Cuando tenemos varios objetos del mismo tipo, `self` le dice a Python **a cuál de ellos** nos estamos refiriendo:

````python
class Producto:
    def __init__(self, nombre: str, precio: float):
        self.nombre: str   = nombre
        self.precio: float = precio

p1 = Producto("Laptop", 1500.0)
p2 = Producto("Mouse",  25.0)
p3 = Producto("Monitor", 800.0)

# Cada objeto tiene su propio espacio en memoria
print(p1.nombre, "→ $", p1.precio)
print(p2.nombre, "→ $", p2.precio)
print(p3.nombre, "→ $", p3.precio)
````

### **Atributos**

Los **atributos** son las variables que pertenecen a un objeto. Se definen en `__init__` usando `self.nombre_atributo = valor`.

Un objeto puede tener tantos atributos como necesitemos, de cualquier tipo: enteros, strings, listas, diccionarios, incluso otros objetos.

````python
class Alumno:
    def __init__(self, nombre: str, legajo: int, notas: list[float] = None):
        self.nombre: str         = nombre
        self.legajo: int         = legajo
        self.notas: list[float]  = notas if notas is not None else []
        self.activo: bool        = True    # atributo con valor fijo inicial

# Con notas
alumno1 = Alumno("Ana", 1001, [8.0, 7.5, 9.0])
print(alumno1.nombre)
print(alumno1.notas)
print(alumno1.activo)

# Sin notas (usa lista vacía por defecto)
alumno2 = Alumno("Bruno", 1002)
print(alumno2.nombre)
print(alumno2.notas)
````

#### Agregar una nota a la lista de un alumno

````python
alumno1 = Alumno("Ana", 1001, [8.0, 7.5])
print("Notas antes:", alumno1.notas)

alumno1.notas.append(9.0)
print("Notas después:", alumno1.notas)
````

### **Métodos**

Los **métodos** son funciones que pertenecen a una clase. Se definen igual que las funciones, pero dentro de la clase y con `self` como primer parámetro.

```python
class MiClase:
    def mi_metodo(self, parametro: tipo) -> tipo_retorno:
        # usa self para acceder a los atributos
        ...
```

Para llamar a un método usamos la notación de punto:
```python
objeto.mi_metodo(argumento)
```

````python
class Alumno:
    def __init__(self, nombre: str, legajo: int, notas: list[float] | None = None):
        self.nombre: str        = nombre
        self.legajo: int        = legajo
        self.notas: list[float] = notas if notas is not None else []

    def calcular_promedio(self) -> float:
        """Calcula y devuelve el promedio de notas del alumno."""
        if len(self.notas) == 0:
            return 0.0
        total: float = 0.0
        for nota in self.notas:
            total = total + nota
        return total / len(self.notas)

    def esta_aprobado(self) -> bool:
        """Devuelve True si el promedio es mayor o igual a 6."""
        return self.calcular_promedio() >= 6.0

    def agregar_nota(self, nota: float) -> None:
        """Agrega una nota a la lista del alumno."""
        self.notas.append(nota)


# Uso
alumno1 = Alumno("Ana", 1001, [8.0, 7.5, 9.0])

print("Promedio:",    alumno1.calcular_promedio())
print("Aprobado:",    alumno1.esta_aprobado())

alumno1.agregar_nota(5.0)
print("Nuevo promedio:", alumno1.calcular_promedio())
````

#### Un método puede recibir parámetros además de `self`

````python
class CuentaBancaria:
    def __init__(self, titular: str, saldo_inicial: float = 0.0):
        self.titular: str  = titular
        self.saldo: float  = saldo_inicial

    def depositar(self, monto: float) -> None:
        """Suma el monto al saldo."""
        self.saldo = self.saldo + monto
        print("Depósito de $", monto, "→ Saldo: $", self.saldo)

    def retirar(self, monto: float) -> bool:
        """Intenta retirar el monto. Devuelve True si hay saldo suficiente."""
        if monto > self.saldo:
            print("Saldo insuficiente. Saldo actual: $", self.saldo)
            return False
        self.saldo = self.saldo - monto
        print("Retiro de $", monto, "→ Saldo: $", self.saldo)
        return True

    def consultar_saldo(self) -> float:
        """Devuelve el saldo actual."""
        return self.saldo


# Uso
cuenta = CuentaBancaria("Ana García", 1000.0)

cuenta.depositar(500.0)
cuenta.retirar(200.0)
cuenta.retirar(2000.0)   # intento fallido
print("Saldo final: $", cuenta.consultar_saldo())
````

### **El método especial `__str__`**

Cuando hacemos `print(objeto)`, Python busca un método llamado `__str__` en la clase. Si no existe, muestra algo poco legible como `<__main__.Alumno object at 0x...>`.

Definiendo `__str__` controlamos exactamente **cómo se representa el objeto como texto**.

```python
def __str__(self) -> str:
    return "texto que representa al objeto"
```

````python
# Sin __str__: representación poco útil
class AlumnoSinStr:
    def __init__(self, nombre: str, legajo: int):
        self.nombre: str = nombre
        self.legajo: int = legajo

a = AlumnoSinStr("Ana", 1001)
print(a)   # algo como: <__main__.AlumnoSinStr object at 0x...>
````

````python
# Con __str__: representación clara y útil
class Alumno:
    def __init__( self, nombre: str, legajo: int, notas: list[float] | None = None):
        self.nombre: str        = nombre
        self.legajo: int        = legajo
        self.notas: list[float] = notas if notas is not None else []

    def calcular_promedio(self) -> float:
        """Calcula el promedio de notas."""
        if len(self.notas) == 0:
            return 0.0
        total: float = 0.0
        for nota in self.notas:
            total = total + nota
        return total / len(self.notas)

    def esta_aprobado(self) -> bool:
        """Devuelve True si el promedio es >= 6."""
        return self.calcular_promedio() >= 6.0

    def agregar_nota(self, nota: float) -> None:
        """Agrega una nota a la lista."""
        self.notas.append(nota)

    def __str__(self) -> str:
        """Representación en texto del alumno."""
        estado: str     = "Aprobado" if self.esta_aprobado() else "Desaprobado"
        promedio: float = round(self.calcular_promedio(), 2)
        return ("Alumno: " + self.nombre +
                " | Legajo: " + str(self.legajo) +
                " | Promedio: " + str(promedio) +
                " | " + estado)


alumno1 = Alumno("Ana",   1001, [8.0, 7.5, 9.0])
alumno2 = Alumno("Bruno", 1002, [5.0, 4.5, 6.0])

print(alumno1)
print(alumno2)
print(type(alumno1), type(alumno2))
````

### **Type hints y docstrings en clases**

Las buenas prácticas que vimos en funciones aplican igual a las clases:

- **Type hints** en todos los atributos y métodos
- **Docstring de clase** justo después de `class NombreClase:`
- **Docstring de método** justo después de `def nombre_metodo(self, ...):`

```python
class MiClase:
    """Descripción de qué representa esta clase."""

    def __init__(self, param: tipo):
        """Inicializa el objeto."""
        self.atributo: tipo = param

    def mi_metodo(self, x: int) -> float:
        """Descripción del método."""
        ...
```

````python
class Producto:
    """Representa un producto en el sistema de ventas."""

    def __init__(self, nombre: str, precio: float, stock: int = 0):
        """Inicializa el producto con nombre, precio y stock.

        Args:
            nombre: nombre del producto
            precio: precio unitario
            stock:  cantidad disponible (por defecto 0)
        """
        self.nombre: str   = nombre
        self.precio: float = precio
        self.stock: int    = stock

    def aplicar_descuento(self, porcentaje: float) -> None:
        """Reduce el precio aplicando un descuento porcentual.

        Args:
            porcentaje: descuento entre 0.0 y 1.0 (ej: 0.20 = 20%)
        """
        self.precio = self.precio * (1 - porcentaje)

    def hay_stock(self) -> bool:
        """Devuelve True si hay al menos una unidad disponible."""
        return self.stock > 0

    def vender(self, cantidad: int) -> bool:
        """Intenta descontar 'cantidad' unidades del stock.

        Returns:
            True si la venta fue exitosa, False si no hay stock suficiente.
        """
        if cantidad > self.stock:
            print("Stock insuficiente. Disponible:", self.stock)
            return False
        self.stock = self.stock - cantidad
        return True

    def __str__(self) -> str:
        """Representación en texto del producto."""
        estado: str = "En stock (" + str(self.stock) + ")" if self.hay_stock() else "Sin stock"
        return self.nombre + " | $" + str(self.precio) + " | " + estado


# Uso
laptop = Producto("Laptop", 1500.0, 10)
print(laptop)

laptop.aplicar_descuento(0.15)    # 15% de descuento
laptop.vender(3)
print(laptop)

# Ver la documentación
help(Producto)
````

### **Listas de objetos**

Una de las combinaciones más poderosas es usar una **lista para guardar múltiples objetos** del mismo tipo. Podemos recorrerla con `for` y llamar a los métodos de cada objeto, exactamente como hicimos con listas de diccionarios.

````python
# Recreamos la clase Alumno completa para este ejemplo
class Alumno:
    """Representa un alumno universitario."""

    def __init__(self, nombre: str, legajo: int, notas: list[float] = None):
        """Inicializa el alumno.

        Args:
            nombre: nombre completo
            legajo: número de legajo único
            notas:  lista de notas (opcional)
        """
        self.nombre: str        = nombre
        self.legajo: int        = legajo
        self.notas: list[float] = notas if notas is not None else []

    def calcular_promedio(self) -> float:
        """Calcula el promedio de notas."""
        if len(self.notas) == 0:
            return 0.0
        total: float = 0.0
        for nota in self.notas:
            total = total + nota
        return total / len(self.notas)

    def esta_aprobado(self) -> bool:
        """Devuelve True si el promedio es >= 6."""
        return self.calcular_promedio() >= 6.0

    def agregar_nota(self, nota: float) -> None:
        """Agrega una nota a la lista."""
        self.notas.append(nota)

    def __str__(self) -> str:
        estado: str     = "Aprobado" if self.esta_aprobado() else "Desaprobado"
        promedio: float = round(self.calcular_promedio(), 2)
        return ("[" + str(self.legajo) + "] " + self.nombre +
                " | Promedio: " + str(promedio) +
                " | " + estado)
````

````python
# Lista de objetos Alumno
curso: list[Alumno] = [
    Alumno("Ana García",   1001, [8.0, 9.0, 7.5]),
    Alumno("Bruno López",  1002, [5.0, 4.5, 6.0]),
    Alumno("Carla Pérez",  1003, [10.0, 9.5, 9.0]),
    Alumno("Diego Martín", 1004, [3.0, 4.0, 5.0]),
    Alumno("Elena Ruiz",   1005, [7.0, 7.5, 8.0]),
]

print("=== LISTADO DEL CURSO ===")
for alumno in curso:
    print(alumno)
````

````python
# Contar aprobados y calcular promedio general del curso
aprobados: int    = 0
total_notas: float = 0.0

for alumno in curso:
    if alumno.esta_aprobado():
        aprobados = aprobados + 1
    total_notas = total_notas + alumno.calcular_promedio()

promedio_curso: float = total_notas / len(curso)

print("Aprobados:", aprobados, "/", len(curso))
print("Promedio del curso:", round(promedio_curso, 2))
````

````python
# Buscar el alumno con mejor promedio
mejor: Alumno = curso[0]

for alumno in curso:
    if alumno.calcular_promedio() > mejor.calcular_promedio():
        mejor = alumno

print("Mejor alumno:", mejor)
````

### **Resumen**

| Concepto | Sintaxis | Descripción |
|---|---|---|
| Definir clase | `class Nombre:` | PascalCase |
| Constructor | `def __init__(self, ...):` | Se ejecuta al crear el objeto |
| Atributo | `self.nombre = valor` | Variable del objeto |
| Método | `def metodo(self, ...):` | Función del objeto |
| Crear objeto | `obj = Clase(args)` | Instanciar la clase |
| Acceder atributo | `obj.atributo` | Leer el valor |
| Llamar método | `obj.metodo(args)` | Ejecutar la función |
| Representación | `def __str__(self) -> str:` | Controla el `print(obj)` |
| Type hint atributo | `self.x: int = valor` | Anotar el tipo del atributo |
| Type hint método | `def m(self, x: int) -> str:` | Anotar parámetros y retorno |
| Docstring de clase | `"""..."""` bajo `class` | Documenta la clase |
| Docstring de método | `"""..."""` bajo `def` | Documenta el método |

````python
# Esqueleto de una clase bien escrita
class MiClase:
    """Descripción de qué representa esta clase."""

    def __init__(self, atributo1: str, atributo2: int):
        """Inicializa el objeto."""
        self.atributo1: str = atributo1
        self.atributo2: int = atributo2

    def mi_metodo(self, param: float) -> float:
        """Descripción del método.

        Args:
            param: descripción

        Returns:
            descripción del retorno
        """
        return param * self.atributo2

    def __str__(self) -> str:
        """Representación en texto."""
        return self.atributo1 + " (" + str(self.atributo2) + ")"


obj = MiClase("ejemplo", 42)
print(obj)
print(obj.mi_metodo(3.0))
````
