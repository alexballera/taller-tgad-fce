### **1.**
**Cree un programa que imprima los siguientes 20 años bisiestos partiendo del año actual**

```python
# Cree un programa que imprima los siguientes 20 años bisiestos partiendo del año actual
import datetime

def es_bisiesto(anio):
    return (anio % 4 == 0 and anio % 100 != 0) or (anio % 400 == 0)

anio_actual = datetime.datetime.now().year
print("Año actual:", anio_actual)

contador = 0
anio = anio_actual

while contador < 20:
    if es_bisiesto(anio):
        print(anio)
        contador += 1
    anio += 1
```

### **2.**
**Cree un programa que devuelva por pantalla todos los números primos menores a 100**

```python
# Cree un programa que devuelva por pantalla todos los números primos menores a 100
def es_primo(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(2, 100):
    if es_primo(num):
        print(num)
```

### **4.**
**Cree una función para ser utilizada en un programa donde el usuario ingrese dos números separados por coma, luego otros dos números seprados por coma, se construya una matriz y devuelva como resultado la matriz multplicada por un escalar ingresado por el usuario**

```python
def matriz_por_escalar():
    fila1 = [float(x.strip()) for x in input("Ingrese dos números separados por coma (fila 1): ").split(",")]
    fila2 = [float(x.strip()) for x in input("Ingrese otros dos números separados por coma (fila 2): ").split(",")]
    escalar = float(input("Ingrese el escalar: "))

    matriz = [fila1, fila2]
    resultado = [[valor * escalar for valor in fila] for fila in matriz]

    print("Matriz original:")
    for fila in matriz:
        print(fila)

    print("\nMatriz multiplicada por el escalar:")
    for fila in resultado:
        print(fila)

    return resultado

matriz_por_escalar()
```

### **6.**
**Cree una función para ser utilizada en un programa, a partir de que el usuario ingrese una secuencia de números y alguno vacío, que devuelva una lista con todos los valores y los vacíos imputados con cero**

#### Mini guía rápida: ¿qué excepción usar?

- `ValueError`: el tipo de dato es correcto, pero el valor no se puede usar.
  - Ejemplo: `int("hola")`, `float("abc")`.
- `TypeError`: se usa un tipo de dato incompatible en una operación.
  - Ejemplo: `"2" + 3`.
- `KeyError`: intentas acceder a una clave que no existe en un diccionario.
  - Ejemplo: `mi_dic["clave_inexistente"]`.
- `IndexError`: intentas acceder a una posición inexistente de una lista/tupla.
  - Ejemplo: `lista[10]` cuando la lista tiene menos elementos.

Recomendación práctica:
1. Captura primero errores específicos (`ValueError`, `KeyError`, etc.).
2. Deja errores genéricos (`Exception`) solo como último recurso.
3. En el `except`, muestra un mensaje claro y decide cómo continuar (por ejemplo, imputar 0 o pedir el dato otra vez).

Documentación oficial de excepciones (español):
https://docs.python.org/es/3.14/library/exceptions.html

```python
def imputar_vacios_con_cero():
    entrada = input("Ingrese una secuencia de números separados por coma (puede haber vacíos): ")
    resultado = []

    for valor in entrada.split(","):
        # strip elimina espacios al inicio y al final para que un valor como "   " quede realmente vacío;
        # así se puede detectar correctamente con valor == "" y, si no está vacío, convertirlo a número.
        valor = valor.strip()
        if valor == "":
            resultado.append(0)
        else:
            # try/except permite intentar convertir el texto a número sin romper el programa.
            # 1) En try usamos float(valor): si valor es válido (por ejemplo "10" o "3.5"), se convierte.
            # 2) Si la conversión funciona, guardamos int si no tiene decimales (10.0 -> 10) o float si los tiene.
            # 3) Si falla (por ejemplo "hola"), Python lanza ValueError.
            # 4) El except captura ese error y, en este ejercicio, imputamos 0 para ese valor inválido.
            # Nota: try/except no devuelve un valor por sí mismo; solo maneja el error y decide cómo continuar.
            try:
                numero = float(valor)
                resultado.append(int(numero) if numero.is_integer() else numero)
            except ValueError:
                print(f"Valor no numérico '{valor}' encontrado. Imputando 0.")
                resultado.append(0)

    return resultado


# Ejemplo de uso
lista_imputada = imputar_vacios_con_cero()
print("Lista con vacíos imputados con 0:", lista_imputada)
```

### **7.**
**Cree una función para ser utilizada en un programa donde a partir de que el usuario ingrese código de producto y cantidad separada por coma como pedido sobre el menú existente, devuelva el detalle del pedido y el total a pagar**

Menu:

1. Empanada de carne    $500 c/u

2. Empanada de jamón y queso    $500 c/u

3. Empanada de humita    $500 c/u

4. Pizza de muzzarela    $15000 c/u

5. Pizza de napolitana    $18000 c/u

6. Pizza de fugazetta    $17000 c/u

```python
def generar_pedido():
    menu = {
        1: ("Empanada de carne", 500),
        2: ("Empanada de jamón y queso", 500),
        3: ("Empanada de humita", 500),
        4: ("Pizza de muzzarela", 15000),
        5: ("Pizza de napolitana", 18000),
        6: ("Pizza de fugazetta", 17000),
    }

    print("Menú:")
    for codigo, (producto, precio) in menu.items():
        print(f"{codigo}. {producto} - ${precio} c/u")

    pedido = {}

    while True:
        entrada = input("Ingrese código,cantidad (Enter vacío para finalizar): ").strip()
        if entrada == "":
            break

        # try/except se usa para manejar errores sin cortar el programa:
        # 1) En try intentamos separar y convertir la entrada a enteros.
        # 2) Si todo sale bien, seguimos validando y guardamos en pedido.
        # 3) Si falla (por ejemplo, no hay coma o escriben texto), Python lanza ValueError.
        # 4) El except captura ese error, muestra un mensaje y el while continúa pidiendo datos.
        # Nota: try/except no "devuelve" un valor por sí mismo; solo controla el flujo ante errores.
        # El valor final de la función se devuelve al final con: return detalle, total.
        try:
            codigo_str, cantidad_str = entrada.split(",")
            codigo = int(codigo_str.strip())
            cantidad = int(cantidad_str.strip())

            if codigo not in menu:
                print("Código inválido.")
                continue
            if cantidad <= 0:
                print("La cantidad debe ser mayor a 0.")
                continue

            pedido[codigo] = pedido.get(codigo, 0) + cantidad

        except ValueError:
            print("Formato inválido. Use: código,cantidad")

    if not pedido:
        print("No se registraron productos.")
        return [], 0

    detalle = []
    total = 0

    print("\nDetalle del pedido:")
    for codigo, cantidad in pedido.items():
        producto, precio = menu[codigo]
        subtotal = cantidad * precio
        detalle.append((codigo, producto, cantidad, precio, subtotal))
        total += subtotal
        print(f"- {producto}: {cantidad} x ${precio} = ${subtotal}")

    print(f"Total a pagar: ${total}")
    return detalle, total


# Ejemplo de uso
detalle_pedido, total_pedido = generar_pedido()
```

### **8.**

**Cree una función para ser utilizada en un programa donde a partir de que el usuario ingrese:**

**- las cantidades y precios consumidas el mes anterior**

**- las cantidades y precios consumidas el mes actual**

**devuelva el Cálculo del Índice de Laspeyre, Paschee y Fisher, donde cada uno son respectivamente**

\begin{align}
         IL_t = {\sum_{k=1}^n q_{i,0}*p_{i;t}\over \sum_{k=1}^n q_{i,0}*p_{i;0}}*100
\end{align}

\begin{align}
         IP_t = {\sum_{k=1}^n q_{i,t}*p_{i;t}\over \sum_{k=1}^n q_{i,t}*p_{i;0}}*100
\end{align}

\begin{align}
         IF_t = \sqrt {IL_t * IP_t}
\end{align}


```python
def calcular_indices_precios():
    # Cantidad de productos
    while True:
        try:
            n = int(input("Ingrese la cantidad de productos: ").strip())
            if n <= 0:
                print("La cantidad debe ser mayor a 0.")
                continue
            break
        except ValueError:
            print("Entrada inválida. Ingrese un número entero.")

    suma_q0_pt = 0.0  # Σ(q0 * pt)
    suma_q0_p0 = 0.0  # Σ(q0 * p0)
    suma_qt_pt = 0.0  # Σ(qt * pt)
    suma_qt_p0 = 0.0  # Σ(qt * p0)

    for i in range(1, n + 1):
        # Mes anterior: q0, p0
        while True:
            try:
                entrada_anterior = input(
                    f"Producto {i} - mes anterior (cantidad,precio): "
                ).strip()
                q0_str, p0_str = entrada_anterior.split(",")
                q0 = float(q0_str.strip())
                p0 = float(p0_str.strip())
                if q0 < 0 or p0 < 0:
                    print("Cantidad y precio deben ser no negativos.")
                    continue
                break
            except ValueError:
                print("Formato inválido. Use: cantidad,precio")

        # Mes actual: qt, pt
        while True:
            try:
                entrada_actual = input(
                    f"Producto {i} - mes actual (cantidad,precio): "
                ).strip()
                qt_str, pt_str = entrada_actual.split(",")
                qt = float(qt_str.strip())
                pt = float(pt_str.strip())
                if qt < 0 or pt < 0:
                    print("Cantidad y precio deben ser no negativos.")
                    continue
                break
            except ValueError:
                print("Formato inválido. Use: cantidad,precio")

        suma_q0_pt += q0 * pt
        suma_q0_p0 += q0 * p0
        suma_qt_pt += qt * pt
        suma_qt_p0 += qt * p0

    if suma_q0_p0 == 0 or suma_qt_p0 == 0:
        print("No se pueden calcular los índices: divisor igual a 0.")
        return None, None, None

    indice_laspeyres = (suma_q0_pt / suma_q0_p0) * 100
    indice_paasche = (suma_qt_pt / suma_qt_p0) * 100
    indice_fisher = (indice_laspeyres * indice_paasche) ** 0.5

    print(f"\nÍndice de Laspeyres (IL_t): {indice_laspeyres:.4f}")
    print(f"Índice de Paasche (IP_t):   {indice_paasche:.4f}")
    print(f"Índice de Fisher (IF_t):    {indice_fisher:.4f}")

    return indice_laspeyres, indice_paasche, indice_fisher


# Ejemplo de uso
il_t, ip_t, if_t = calcular_indices_precios()
```
