## Actividad 4

En un comercio se vende los productos: A,B,C,D,E,F,G,H uyos precios sin IVA son: 10, 15, 20, 25, 30, 35, 40, 45.


Cree una función que ejecute el siguiente programa:


Si el cliente paga con débito, obtiene 15% de descuento, si en cambio paga con tarjeta de crédito obtiene un 20% de descuento. Si paga en efectivo, no obtiene descuento y si paga con billetera virtual obtiene un 25% de descuento 


Finalmente que devuelva lo que el cliente debe pagar después de aplicar el descuento, si corresponde.


Pruebe que su programa funcione correctamente, siendo que el cliente debe ingresar el/los producto/s comprado/s en mayúscula, las cantidades en el mismo orden que los productos y el medio de pago en mayúscula

```python
# Actividad 4
def calcular_total_compra(productos_ingresados, cantidades_ingresadas, medio_pago):
    """
    Calcula el total a pagar según productos, cantidades y medio de pago.

    Parámetros:
      productos_ingresados: str con productos separados por coma (ej: "A,C,H")
      cantidades_ingresadas: str con cantidades separadas por coma (ej: "2,1,3")
      medio_pago: str en mayúscula (DEBITO, TARJETA DE CREDITO, EFECTIVO, BILLETERA VIRTUAL)

    Devuelve:
      dict con detalle de compra, subtotal, descuento y total.
    """
    precios = {
        "A": 10, "B": 15, "C": 20, "D": 25,
        "E": 30, "F": 35, "G": 40, "H": 45
    }

    descuentos = {
        "DEBITO": 0.15,
        "TARJETA DE CREDITO": 0.20,
        "EFECTIVO": 0.00,
        "BILLETERA VIRTUAL": 0.25,
    }

    productos = [p.strip().upper() for p in productos_ingresados.split(",") if p.strip()]
    cantidades = [c.strip() for c in cantidades_ingresadas.split(",") if c.strip()]
    medio_pago = medio_pago.strip().upper()

    if len(productos) == 0:
        raise ValueError("Debe ingresar al menos un producto.")
    if len(productos) != len(cantidades):
        raise ValueError("La cantidad de productos y cantidades debe coincidir.")
    if medio_pago not in descuentos:
        raise ValueError("Medio de pago inválido.")

    cantidades_int = []
    for c in cantidades:
        if not c.isdigit():
            raise ValueError("Las cantidades deben ser números enteros no negativos.")
        cantidades_int.append(int(c))

    detalle = []
    subtotal = 0.0

    for producto, cantidad in zip(productos, cantidades_int):
        if producto not in precios:
            raise ValueError(f"Producto inválido: {producto}")
        importe = precios[producto] * cantidad
        subtotal += importe
        detalle.append({
            "producto": producto,
            "cantidad": cantidad,
            "precio_unitario": precios[producto],
            "importe": importe,
        })

    porcentaje_descuento = descuentos[medio_pago]
    descuento = subtotal * porcentaje_descuento
    total = subtotal - descuento

    return {
        "detalle": detalle,
        "subtotal": round(subtotal, 2),
        "porcentaje_descuento": porcentaje_descuento,
        "descuento": round(descuento, 2),
        "total": round(total, 2),
        "medio_pago": medio_pago,
    }


def mostrar_resultado(resultado):
    print("\nDetalle de la compra:")
    for item in resultado["detalle"]:
        print(
            f"Producto {item['producto']}: {item['cantidad']} unidad(es) x "
            f"${item['precio_unitario']} = ${item['importe']:.2f}"
        )
    print(f"Subtotal: ${resultado['subtotal']:.2f}")
    print(
        f"Descuento ({resultado['porcentaje_descuento']*100:.0f}% - {resultado['medio_pago']}): "
        f"${resultado['descuento']:.2f}"
    )
    print(f"Total a pagar: ${resultado['total']:.2f}")


# Programa interactivo
productos_txt = input("Ingrese el/los producto/s en mayúscula separados por coma (ej: A,C,H): ")
cantidades_txt = input("Ingrese las cantidades en el mismo orden separadas por coma (ej: 2,1,3): ")
medio_pago_txt = input(
    "Ingrese medio de pago en mayúscula (DEBITO, TARJETA DE CREDITO, EFECTIVO, BILLETERA VIRTUAL): "
)

resultado_compra = calcular_total_compra(productos_txt, cantidades_txt, medio_pago_txt)
mostrar_resultado(resultado_compra)

# Pruebas rápidas
print("\nPruebas rápidas:")
print(calcular_total_compra("A,B", "1,2", "DEBITO"))
print(calcular_total_compra("H", "3", "EFECTIVO"))
```
