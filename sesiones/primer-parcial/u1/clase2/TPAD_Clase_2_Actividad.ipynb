## Ejercicio 1
1- A partir de que el usuario ingrese su nombre y apellido, su nacionalidad y su edad, conforme una lista. El nombre, el apellido y la nacionalidad conviértalo a mayúscula. Con la edad calcule el año de nacimiento. Finalmente, que el programa devuelva “El ciudadano NOMBRE y APELLIDO de nacionalidad.
 
NACIONALIDAD nació en el año AÑO”

```python
# Ejercicio 1
from datetime import datetime

nombre: str = input("Ingresá tu nombre: ").strip()
apellido: str = input("Ingresá tu apellido: ").strip()
nacionalidad: str = input("Ingresá tu nacionalidad: ").strip()
edad: int = int(input("Ingresá tu edad: "))

datos_ciudadano: list = [nombre.upper(), apellido.upper(), nacionalidad.upper(), edad]

anio_actual: int = datetime.now().year
anio_nacimiento: int = anio_actual - datos_ciudadano[3]

print(
    f"El ciudadano {datos_ciudadano[0]} y {datos_ciudadano[1]} "
    f"de nacionalidad {datos_ciudadano[2]} nació en el año {anio_nacimiento}"
)
```

## Ejercicio 2

2- Cree una lista con los siguientes productos: A, B, C, D, E, F. Cree otra lista con los precios 10, 25, 38, 50, 62, 75. Ahora almacene el nombre de cada producto por un lado y el valor de cada cantidad comprada por otro, que ingrese el usuario de tres de los productos. Finalmente, que el programa devuelva “El cliente compró X unidades de X a $X, Y unidades de Y a $Y, Z unidades de Z a $Z y en total gastó $MONTO”

```python
# Ejercicio 2
productos: list[str] = ["A", "B", "C", "D", "E", "F"]
precios: list[float] = [10, 25, 38, 50, 62, 75]

print("Productos disponibles:", productos)
print("Precios:", precios)

productos_comprados: list[str] = []
subtotales: list[float] = []
cantidades: list[int] = []

for i in range(3):
    producto: str = input(f"Producto {i + 1}: ").strip().upper()
    cantidad: int = int(input(f"Cantidad de {producto}: "))

    if producto not in productos:
        print(f"Producto inválido: {producto}")
        break

    indice: int = productos.index(producto)
    precio_unitario: float = precios[indice]
    subtotal: float = cantidad * precio_unitario

    productos_comprados.append(producto)
    cantidades.append(cantidad)
    subtotales.append(subtotal)

if len(productos_comprados) == 3:
    monto_total: float = sum(subtotales)
    print(
        f"El cliente compró {cantidades[0]} unidades de {productos_comprados[0]} a ${subtotales[0]:.2f}, "
        f"{cantidades[1]} unidades de {productos_comprados[1]} a ${subtotales[1]:.2f}, "
        f"{cantidades[2]} unidades de {productos_comprados[2]} a ${subtotales[2]:.2f} "
        f"y en total gastó ${monto_total:.2f}"
    )
```
