Si quisieramos operar sobre cada elemento de la lista anidada (en este caso una matriz numérica) deberemos recorrerlo (se debe utilizar un FOR) y sobre cada elemento aplicar una operación. Esto lo veremos la próxima clase.

**En base a las distintas funciones que utilizamos sobre una lista, observamos que hay de dos tipos:**

**Las que NO modifican la lista**

```python
len(lista_all) #Devuelve el número de elementos de la lista.
```

```python
min(lista_all) # Devuelve el mínimo elemento de la lista siempre que los datos sean comparables.
```

```python
max(lista_all) # Devuelve el máximo elemento de la lista siempre que los datos sean comparables.
```

```python
sum(lista_all[0])  # suma los elementos de una sublista; lista_all es una lista de listas
```

```python
sum(lista_all[1]) #entonces debo sumar los elementos del cada elemento en forma individual porque no es posible sumar listas como si fuesen elementos
```

```python
[1,2,3] in lista_all #Devuelve True si el elemento lista indicado pertenece a la lista y False en caso contrario.
```

```python
lista_all.index([1,2,3]) # Devuelve la posición que ocupa en la lista el primer elemento con valor [1,2,3].
```

```python
lista_all.count([1,2,3]) # Devuelve el número de veces que el valor [1,2,3] está contenido en la lista.
```

```python
all(lista_all) # Devuelve True si todos los elementos de la lista son True y False en caso contrario.
```

```python
any(lista_all) # Devuelve True si algún elemento de la lista es True y False en caso contrario.
```

**Las que SI modifican la lista**

```python
lista_1 + lista_2 # Crea una nueva lista concatenan los elementos de la listas l1 y l2.
```

```python
lista_1.append(4) # Añade dato al final de la lista l.
lista_1 #observo que se modificó
```

```python
lista_1.extend(lista_2) # Añade los datos de lista 2 al final de la lista l.
lista_1
```

```python
lista_1.insert(4, None) # Inserta None (vacío) en la posición indicada de la lista l y desplaza los elementos una posición a partir de la posición indicada.
lista_1
```

```python
lista_1.remove(None) # Elimina el primer elemento con valor None en la lista l y desplaza los que están por detrás de él una posición hacia delante.
lista_1
```

```python
lista_1.pop(4) # Devuelve el dato de la posición 4 y lo elimina de la lista l, desplazando los elementos por detrás de él una posición hacia delante.
lista_1
```

```python
lista_1.reverse() #invierte el orden de los elementos de la lista l.
lista_1
```

```python
lista_1.sort() #Ordena los elementos de la lista l de acuerdo al orden predefinido, siempre que los elementos sean comparables.
lista_1
```

**Ejemplos**

A partir de que un usuario ingrese los valores 1,2,3, cree una lista con ellos. Utilizando la función append() agregue el valor 4 y 4+1 a la lista. Finalmente, muestre por pantalla "La lista en orden inverso es: lista", donde lista debe mostrarse en orden inverso.

```python
lista=[input(),input(),input()]
lista.append(4)
lista.append(4+1)
lista.reverse()
print('La lista en orden inverso es: '+str(lista))
```

Cree una lista con los productos computadora, teclado, software. Ordene la lista alfabéticamente. Obtenga el índice de correspondiente al elemento software. Finalmente, imprima "La posición del elemento software en la lista es: c" donde c es el valor de la posición

```python
productos=['computadora', 'teclado', 'software']
productos.sort()
print('La posición del elemento software en la lista es: '+str(productos.index('software')))
```

## **2.Tuples**


Las tuplas también son un objeto secuencial. Pero a diferencia de las listas no se pueden mutar sus elementos ni añadirle nuevos.
Para crear una tupla utilizamos paréntesis y separamos los elementos con coma.

```python
tupla=(1,2,'tres')
```

```python
type(tupla)
```

```python
tupla
```

```python
len(tupla) #vemos cuantos elementos tiene la tupla
```

```python
tupla[0] #seleccionamos un elemento determinado
```

```python
tupla[1:]
```

```python
# como mencionamos al comienzo las tuplas son inmutables, es decir, no pueden modificarse sus elementos
tupla(0)='a'
```

```python
tupla.append('a')
```
