## Actividad 3

Escriba un programa donde:

  • el usuario ingrese 3 las siguientes tres fechas de nacimiento: 26/2/1983, 12/6/1993, 18/11/2003 en formato Año/Mes/día para conformar un vector de precios solo con el valor del año

  • el usuario ingrese el nombre de 3 productos. Conforme un diccionario donde sus valores sean las cantidades compradas, siendo estas ultimas la edad calculada. Luego conforme un vector únicamente de cantidades.

  • Con los vectores de cantidades y precios conforme una matriz, agregue una fila del subtotal, otra fila aplicando un descuento del 20% y una fila de total

  • Finalmente que el programa devuelva lo gastado por el usuario después de aplicar un descuento del 20% con la siguiente especificación: El usuario compró: X unidades del producto X que suma $X, Y unidades del producto Y que suma $Y, Z unidades del producto Z que suma $Z, siendo un subtotal de $S, descuento del 20% $D, Total gastado $T redondeado a dos decimales

```python
# Actividad 3
from datetime import datetime, date

def parsear_fecha(texto):
    """Acepta formatos AAAA/MM/DD o DD/MM/AAAA."""
    texto = texto.strip().replace("-", "/")
    partes = texto.split("/")
    if len(partes) != 3:
        raise ValueError("Formato inválido. Use AAAA/MM/DD o DD/MM/AAAA.")

    a, b, c = partes
    if len(a) == 4:  # AAAA/MM/DD
        anio, mes, dia = int(a), int(b), int(c)
    else:            # DD/MM/AAAA
        dia, mes, anio = int(a), int(b), int(c)

    return date(anio, mes, dia)

def calcular_edad(fecha_nacimiento, hoy=None):
    if hoy is None:
        hoy = date.today()
    edad = hoy.year - fecha_nacimiento.year
    if (hoy.month, hoy.day) < (fecha_nacimiento.month, fecha_nacimiento.day):
        edad -= 1
    return edad

print("Ingrese 3 fechas de nacimiento (AAAA/MM/DD o DD/MM/AAAA):")
fechas_nacimiento = []
for i in range(3):
    while True:
        try:
            entrada = input(f"Fecha {i+1}: ")
            fecha = parsear_fecha(entrada)
            fechas_nacimiento.append(fecha)
            break
        except ValueError as e:
            print(f"Error: {e}")

# Vector de precios: solo año de cada fecha
precios = [f.year for f in fechas_nacimiento]

print("\nIngrese el nombre de 3 productos:")
productos = [input(f"Producto {i+1}: ").strip() for i in range(3)]

# Cantidades compradas = edades calculadas
cantidades = [calcular_edad(f) for f in fechas_nacimiento]
productos_cantidades = dict(zip(productos, cantidades))

# Matriz base: [cantidad, precio, total_por_producto]
matriz = []
for cantidad, precio in zip(cantidades, precios):
    matriz.append([cantidad, precio, cantidad * precio])

subtotal = sum(fila[2] for fila in matriz)
descuento = subtotal * 0.20
total = subtotal - descuento

# Agregar filas de subtotal, descuento y total
matriz.append(["-", "SUBTOTAL", subtotal])
matriz.append(["-", "DESCUENTO 20%", -descuento])
matriz.append(["-", "TOTAL", total])

print("\nDiccionario producto -> cantidad:")
print(productos_cantidades)

print("\nVector de cantidades:")
print(cantidades)

print("\nVector de precios (años):")
print(precios)

print("\nMatriz final [cantidad, precio/etiqueta, importe]:")
for fila in matriz:
    print(fila)

sumas_por_producto = [cantidades[i] * precios[i] for i in range(3)]
print(
    f"\nEl usuario compró: "
    f"{cantidades[0]} unidades del producto {productos[0]} que suma ${sumas_por_producto[0]:.2f}, "
    f"{cantidades[1]} unidades del producto {productos[1]} que suma ${sumas_por_producto[1]:.2f}, "
    f"{cantidades[2]} unidades del producto {productos[2]} que suma ${sumas_por_producto[2]:.2f}, "
    f"siendo un subtotal de ${subtotal:.2f}, descuento del 20% ${descuento:.2f}, "
    f"Total gastado ${total:.2f}"
)
```
