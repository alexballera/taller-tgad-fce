### **Ejercicio 2** — Operaciones aritméticas básicas

Una empresa vendió **347 unidades** de un producto a **$2.850,00** cada una. Sobre el precio unitario se aplica un **IVA del 21%**.

Calculá y mostrá:
1. El ingreso bruto (sin IVA)
2. El IVA total recaudado
3. El ingreso total (con IVA)
4. El precio unitario con IVA

Usá type hints en todas las variables.

````python
# Tu resolución acá
````

#### Solución — Ejercicio 2

````python
unidades: int          = 347
precio_unitario: float = 2850.0
tasa_iva: float        = 0.21

ingreso_bruto: float   = unidades * precio_unitario
iva_total: float       = ingreso_bruto * tasa_iva
ingreso_total: float   = ingreso_bruto + iva_total
precio_con_iva: float  = precio_unitario * (1 + tasa_iva)

print("Ingreso bruto:         $", ingreso_bruto)
print("IVA total:             $", iva_total)
print("Ingreso total:         $", ingreso_total)
print("Precio unitario + IVA: $", precio_con_iva)
````

### **Ejercicio 3** — División entera y resto (`//` y `%`)

Un depósito recibió **1.250 unidades** de un producto. Cada caja contiene **48 unidades**.

Calculá y mostrá:
1. Cuántas cajas completas se pueden armar
2. Cuántas unidades sobran (no caben en ninguna caja)
3. Si sobran unidades, cuántas unidades faltan para completar una caja más

La salida esperada:
```
Unidades totales: 1250
Cajas completas: 26
Unidades sobrantes: 2
Faltan 46 unidades para completar otra caja
```

````python
# Tu resolución acá
````

#### Solución — Ejercicio 3

````python
unidades: int        = 1250
por_caja: int        = 48

cajas_completas: int = unidades // por_caja
sobrantes: int       = unidades % por_caja
faltan: int          = por_caja - sobrantes

print("Unidades totales:", unidades)
print("Cajas completas:",  cajas_completas)
print("Unidades sobrantes:", sobrantes)
print("Faltan", faltan, "unidades para completar otra caja")
````

### **Ejercicio 4** — Indexing y slicing de strings

Dada la siguiente variable:

```python
codigo: str = "UCA-AN-2024-G01"
```

Este código tiene el formato `INSTITUCION-CARRERA-AÑO-GRUPO`. Sin usar métodos como `.split()`, usá **indexing y slicing** para extraer:

1. La institución (`"UCA"`) → primeros 3 caracteres
2. La carrera (`"AN"`) → caracteres 4 y 5
3. El año (`"2024"`) → caracteres 7 al 10
4. El grupo (`"G01"`) → últimos 3 caracteres
5. El código completo al revés
6. Solo los caracteres en posiciones pares (0, 2, 4, ...)

Mostrá cada resultado con una etiqueta descriptiva.

````python
# Tu resolución acá
````

#### Solución — Ejercicio 4

````python
codigo: str = "UCA-AN-2024-G01"

institucion: str = codigo[:3]
carrera: str     = codigo[4:6]
anio: str        = codigo[7:11]
grupo: str       = codigo[-3:]
al_reves: str    = codigo[::-1]
pares: str       = codigo[::2]

print("Institución:", institucion)
print("Carrera:    ", carrera)
print("Año:        ", anio)
print("Grupo:      ", grupo)
print("Al revés:   ", al_reves)
print("Pos. pares: ", pares)
````

### **Ejercicio 6** — f-strings y formato de salida

Tenés los siguientes datos de un producto:

```python
nombre: str    = "Monitor 27 pulgadas"
precio: float  = 87650.0
stock: int     = 14
descuento: float = 0.12
```

Usando **exclusivamente f-strings**, generá la siguiente salida. Prestá atención al formato de los números:

```
Producto:        Monitor 27 pulgadas
Precio original: $87650.00
Descuento:       12.00%
Ahorro:          $10518.00
Precio final:    $77132.00

```

````python
# Tu resolución acá
````

#### Solución — Ejercicio 6

````python
nombre: str      = "Monitor 27 pulgadas"
precio: float    = 87650.0
stock: int       = 14
descuento: float = 0.12

ahorro: float       = precio * descuento
precio_final: float = precio - ahorro

print(f"Producto:        {nombre}")
print(f"Precio original: ${precio:.2f}")
print(f"Descuento:       {descuento * 100:.2f}%")
print(f"Ahorro:          ${ahorro:.2f}")
print(f"Precio final:    ${precio_final:.2f}")
print(f"Stock:           {stock} unidades")
````

### **Ejercicio 7** — Conversión de tipos

El siguiente programa simula la lectura de datos ingresados por un usuario. Los datos llegan siempre como **strings**:

```python
entrada_precio    = "1250.50"
entrada_cantidad  = "8"
entrada_descuento = "15"
```

Convertí cada variable al tipo adecuado y calculá:

1. El subtotal (precio × cantidad)
2. El monto del descuento
3. El total final
4. Un mensaje de resumen con f-string:
```
Resumen: 8 unidades a $1250.50 c/u | Descuento: 15% | Total: $8503.40
```

````python
# Tu resolución acá
````

#### Solución — Ejercicio 7

````python
entrada_precio:    str = "1250.50"
entrada_cantidad:  str = "8"
entrada_descuento: str = "15"

precio: float    = float(entrada_precio)
cantidad: int    = int(entrada_cantidad)
descuento: float = float(entrada_descuento)

subtotal: float        = precio * cantidad
monto_descuento: float = subtotal * (descuento / 100)
total: float           = subtotal - monto_descuento

print(f"Resumen: {cantidad} unidades a ${precio} c/u | "
      f"Descuento: {int(descuento)}% | Total: ${total:.2f}")
````

### **Ejercicio 8** — `input()` y conversión de tipos

Escribí un programa que le pida al usuario:

1. Su nombre
2. Su salario mensual bruto (como número)
3. Su porcentaje de ahorro mensual (como número entre 0 y 100)

Y luego muestre:

```
Hola, Martín!
Tu salario bruto es: $45000.00
Ahorrás el 20% → $9000.00 por mes
En un año acumulás: $108000.00
```

> Recordá que `input()` siempre devuelve un string: convertí los números antes de operar.

````python
# Tu resolución acá
````

#### Solución — Ejercicio 8

````python
nombre: str             = input("Tu nombre: ")
salario_str: str        = input("Tu salario mensual bruto: ")
porcentaje_str: str     = input("Tu porcentaje de ahorro (0-100): ")

salario: float          = float(salario_str)
porcentaje: float       = float(porcentaje_str)

ahorro_mensual: float   = salario * (porcentaje / 100)
ahorro_anual: float     = ahorro_mensual * 12

print(f"Hola, {nombre}!")
print(f"Tu salario bruto es: ${salario:.2f}")
print(f"Ahorrás el {int(porcentaje)}% → ${ahorro_mensual:.2f} por mes")
print(f"En un año acumulás: ${ahorro_anual:.2f}")
````

### **Ejercicio 9** — Manipulación avanzada de strings

Escribí un programa que procese el nombre completo de una persona. El usuario ingresa su nombre y apellido **en cualquier combinación de mayúsculas y minúsculas** (por ejemplo: `"ana GARCÍA"`, `"BRUNO lopez"`, `"carla Pérez"`).

El programa debe:

1. Leer el nombre completo con `input()`
2. Normalizar el formato a **título** (primera letra de cada palabra en mayúscula)
3. Separar nombre y apellido (asumí que son exactamente dos palabras)
4. Generar un nombre de usuario con el formato: primera letra del nombre + apellido en minúsculas
5. Mostrar el resultado:

```
Nombre completo:  Ana García
Nombre:           Ana
Apellido:         García
Usuario:          agarcia
Largo del nombre: 9 caracteres (sin espacios)
```

**Pistas:** `.title()`, `.split()`, indexing de listas, `.replace()`

````python
# Tu resolución acá
````

#### Solución — Ejercicio 9

````python
entrada: str          = input("Ingresá tu nombre completo: ")

nombre_completo: str  = entrada.strip().title()
partes: list          = nombre_completo.split()
nombre: str           = partes[0]
apellido: str         = partes[1]
usuario: str          = nombre[0].lower() + apellido.lower()
largo: int            = len(nombre_completo.replace(" ", ""))

print(f"Nombre completo:  {nombre_completo}")
print(f"Nombre:           {nombre}")
print(f"Apellido:         {apellido}")
print(f"Usuario:          {usuario}")
print(f"Largo del nombre: {largo} caracteres (sin espacios)")
````

### **Ejercicio 10** — Integrador

Un empleado tiene los siguientes datos:

```
Nombre:                     "María José Rodríguez"
Salario bruto:              95000.0
Años de antigüedad:         7
Adicional por antigüedad:   2% por año
Impuesto a las ganancias:   17%
```

Calculá y mostrá un **recibo de sueldo** con el siguiente formato:

```
========================================
           RECIBO DE SUELDO
========================================
Empleado:            MARÍA JOSÉ RODRÍGUEZ
Iniciales:           MJR
----------------------------------------
Salario base:        $95000.00
Adicional (7 años):  $13300.00
                     ----------
Salario bruto:       $108300.00
Descuento (17%):    -$18411.00
                     ----------
SALARIO NETO:        $89889.00
========================================
```

**Además:**
- Las iniciales se obtienen tomando la primera letra de cada palabra del nombre
- El nombre debe mostrarse en mayúsculas
- Usá f-strings con `:.2f` para todos los montos

````python
# Tu resolución acá
````

#### Solución — Ejercicio 10

````python
nombre: str            = "María José Rodríguez"
salario_base: float    = 95000.0
anios: int             = 7
tasa_antiguedad: float = 0.02
tasa_impuesto: float   = 0.17

# Cálculos
adicional: float       = salario_base * anios * tasa_antiguedad
salario_bruto: float   = salario_base + adicional
descuento: float       = salario_bruto * tasa_impuesto
salario_neto: float    = salario_bruto - descuento

# Iniciales: primera letra de cada palabra
iniciales: str = ""
for palabra in nombre.split():
    iniciales = iniciales + palabra[0].upper()

# Formato de salida
sep1: str = "=" * 40
sep2: str = "-" * 40

print(sep1)
print("           RECIBO DE SUELDO")
print(sep1)
print(f"Empleado:            {nombre.upper()}")
print(f"Iniciales:           {iniciales}")
print(sep2)
print(f"Salario base:        ${salario_base:.2f}")
print(f"Adicional ({anios} años):  ${adicional:.2f}")
print(f"                     ----------")
print(f"Salario bruto:       ${salario_bruto:.2f}")
print(f"Descuento ({int(tasa_impuesto*100)}%):    -${descuento:.2f}")
print(f"                     ----------")
print(f"SALARIO NETO:        ${salario_neto:.2f}")
print(sep1)
````
