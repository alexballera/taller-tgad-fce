####**Muestreo aletorio estratificado**

Si recordamos de estadística:

image.png

```python
data1=pd.DataFrame(np.repeat(1,150))
data2=pd.DataFrame(np.repeat(2,250))
data3=pd.DataFrame(np.repeat(3,300))
data4=pd.DataFrame(np.repeat(4,200))
data5=pd.DataFrame(np.repeat(5,100))

data=pd.concat([data1,data2,data3,data4,data5])
data.columns=['Grupo']
data['Individuos']=data['Grupo']
data
```

```python
#se crea una muestra de tamaño n=5, estratificada por la columna Grupo
muestra=data.groupby('Grupo', group_keys=False).apply(lambda x: x.sample(frac=0.1))
muestra
```

Lo verificamos:



```python
#cantidades que debemos tener por estrato
round((100/1000)*data.groupby(data['Individuos']).count(),0)
```

```python
#observamos cuantos datos hay por estrato en la muestra generada al inicio

muestra.groupby(muestra['Grupo']).count()
```

Para el DataFrame, por ejemplo consideremos que el estrato viene dado por la variable occupation_status

Primero necesitamos convertir a dicha variable en categorías numéricas

```python
DATA_1=DATA.copy() #para no perder el dataframe original trabajamos con una copia

DATA_1['occupation_status']=DATA_1['occupation_status'].astype('category').cat.codes

DATA_1
```

Si nuestro universo fuese este dataframe y los estratos occupation_status, y queremso una muestra de 100 casos, entonces

```python
#tenemos 50000 casos => una muestra tamaño 100 es el 0.2%
DATA_1.shape
```

```python
#se crea una muestra de tamaño n=100, estratificada por la columna occupation_status
muestra_1=DATA_1.groupby('occupation_status', group_keys=False).apply(lambda x: x.sample(frac=0.002))
muestra_1
```

```python
muestra_1.groupby(muestra_1['occupation_status']).count()
```

Verificamos

Observamos que tenemos cantidad de casos correcta en la muestra generada

####**Muestreo aleatorio por conglomerado**

Si recordamos de Estadística, este método era:

image.png

```python
# Número de conglomerados a seleccionar
n_conglomerados = 3

# Seleccionar conglomerados aleatorios
conglomerados_seleccionados = np.random.choice(DATA_1['occupation_status'].unique(), n_conglomerados, replace=False)

# Fijar cantidad de elementos dentro de cada conglomerado
n_elementos_por_conglomerado = 30
# Por ejemplo, quiero una muestra de 90 elementos y especifiqué 3 conglomerados, quiero que haya 30 elementos por conglomerado

# Crear un DataFrame para almacenar las filas seleccionadas
muestra_2_cong = pd.DataFrame()

# Para cada conglomerado seleccionado
for conglomerado in conglomerados_seleccionados:
    # Filtrar los elementos del conglomerado
    df_conglomerado = DATA_1[DATA_1['occupation_status'] == conglomerado]
    # Seleccionar una muestra aleatoria con reemplazo dentro del conglomerado
    muestra_conglomerado = df_conglomerado.sample(n=n_elementos_por_conglomerado, random_state=2026, replace=True)
    # Agregar la muestra del conglomerado a la muestra total
    muestra_2_cong = pd.concat([muestra_2_cong, muestra_conglomerado])

# Observar la muestra final generada
muestra_2_cong
```

```python
muestra_2_cong=muestra_2_cong.reset_index(drop=True)
muestra_2_cong.groupby('occupation_status').count()
```

Se observan exactamente 30 casos por conglomerado

### **Unir DataFrames**

```python
#se importan los datos a través de la API, teniendo en cuenta el formato que finalmente se necesitará según lo que se requiera hacer.

#importación de librerías

import yfinance as yf

Datos_MSFT = pd.DataFrame(yf.download('MSFT', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_MSFT=Datos_MSFT.assign(SYMBOL_YAHOO_FINANCE='MSFT')
Datos_MSFT.columns=Datos_MSFT.columns.droplevel(1)

Datos_AAPL = pd.DataFrame(yf.download('AAPL', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_AAPL=Datos_AAPL.assign(SYMBOL_YAHOO_FINANCE='AAPL')
Datos_AAPL.columns=Datos_AAPL.columns.droplevel(1)

Datos_IBM = pd.DataFrame(yf.download('IBM', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_IBM=Datos_IBM.assign(SYMBOL_YAHOO_FINANCE='IBM')
Datos_IBM.columns=Datos_IBM.columns.droplevel(1)


#GENERACIÓN DE UN SOLO DATAFRAME
DATA_YAHOO=pd.concat([Datos_MSFT,Datos_AAPL,Datos_IBM])
#RESETEO DE INDICE
DATA_YAHOO.reset_index(drop=True, inplace=True)
```

```python
DATA_YAHOO
```

```python
#Creamos un dataframe con datos adicionales (podría importarlo desde algún otra fuente origen)

#leemos el csv importado
EMPRESAS=pd.DataFrame({'NAME_COMPANY':['Microsoft Corp.', 'Apple Inc.', 'International Business Machines Corp.'],
                       'SYMBOL_YAHOO_FINANCE':['MSFT', 'AAPL','IBM'],
                       'COUNTRY':'US',
                       'STATE':['New Mexico', 'California', 'New York'],
                       'CITY':['Albuquerque','Los Altos','Endicott'],
                       'YEAR_FUNDATION':[1975,1976,1911],
                       'SECTOR':'Tecnologia'})
EMPRESAS
```

###**Agregar filas a un DataFrame**

Agregar filas es lo que se conoce como concatenar

Para esto será necesario que las columnas coincidan, incluso en orden

```python
#se importan los datos a través de la API, teniendo en cuenta el formato que finalmente se necesitará según lo que se requiera hacer.

Datos_MSFT = pd.DataFrame(yf.download('MSFT', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_MSFT=Datos_MSFT.assign(SYMBOL_YAHOO_FINANCE='MSFT')
Datos_MSFT.columns=Datos_MSFT.columns.droplevel(1)

Datos_AAPL = pd.DataFrame(yf.download('AAPL', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_AAPL=Datos_AAPL.assign(SYMBOL_YAHOO_FINANCE='AAPL')
Datos_AAPL.columns=Datos_AAPL.columns.droplevel(1)

Datos_IBM = pd.DataFrame(yf.download('IBM', start='2023-03-20', end='2023-03-25')).reset_index()
Datos_IBM=Datos_IBM.assign(SYMBOL_YAHOO_FINANCE='IBM')
Datos_IBM.columns=Datos_IBM.columns.droplevel(1)


#GENERACIÓN DE UN SOLO DATAFRAME
DATA_YAHOO_1=pd.concat([Datos_MSFT,Datos_AAPL,Datos_IBM])
#RESETEO DE INDICE
DATA_YAHOO_1.reset_index(drop=True, inplace=True)
```
