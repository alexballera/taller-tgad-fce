#### Apply

```python
df["Rating"]
```

```python
def rating_cat(rating):
  level="Bajo"
  if rating>=2 and rating<4:
    level="Medio"
  elif rating>=4:
    level="Alto"

  return level

df["Rating_category"]=df["Rating"].apply(rating_cat)
df.head(10)
```

```python
def rating_cat_fila(fila):
  level="Bajo"
  if fila["Rating"]>=2 and fila["Rating"]<4:
    level="Medio"
  elif fila["Rating"]>=4:
    level="Alto"

  return level

df["Rating_category"]=df.apply(rating_cat_fila,axis="columns")
df.head(10)
```

```python
def salary_porc(fila):
  if fila["Company_Name"]=="Sasken":
    porc_salary=fila["Salary"]*0.01
  else:
    porc_salary=fila["Salary"]*0.21

  return porc_salary

df["Salary_porc"]=df.apply(salary_porc,axis="columns")
df.head(10)
```

```python
df["Rating_category_lambda"]=df["Rating"].apply(lambda rating: "Bajo" if rating<2 else ("Medio" if (rating>=2 and rating<4) else "Alto") )
df.head(10)
```

# Unión de DataFrames

https://pandas.pydata.org/docs/user_guide/merging.html

Concat



```python
d1 = {"Nombre": ["Pedro", "Juan"],
      "id": [1, 2]}
d2 = {"Nombre": "Mariano",
      "id": 3}

df1 = pd.DataFrame(d1, index= [1, 2])
df2 = pd.DataFrame(d2, index=[3])

df3 = pd.concat([df1, df2])

df3
```

```python
df1.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
d1 = {"Nombre": ["Pedro", "Juan"],
      "id": [1, 2]}
d2 = {"Profesión": ["Médico", "Ingeniero"]}

df1 = pd.DataFrame(d1, index=[1, 2])
df2 = pd.DataFrame(d2, index=[1, 2])

df3 = pd.concat([df1, df2], axis=1)
df3
```

Merge

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo"],
      "id": [1, 2, 3],
      "id_profesion": [1,2,1]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2,3])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2)
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2, 3, 4])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2,how='inner')
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2, 3, 4])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2,how='outer')
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2, 3, 4])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2,how='left')
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2, 3, 4])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2,how='right')
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2, 3, 4])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2,how='right',left_on="id_profesion",right_on="id_profesion")
result
```
