* Define nuevas estructuras de datos basadas en los arrays de la librería NumPy pero con nuevas funcionalidades.

Merge

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo"],
      "id": [1, 2, 3],
      "id_profesion": [1,2,1]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index=[1, 2,3])
df2 = pd.DataFrame(d2, index=[1, 2])

result = pd.merge(df1,df2)
result
```

```python
d1 = {"Nombre": ["Pedro", "Juan","Pablo","Manuel"],
      "id": [1, 2, 3, 4],
      "id_profesion": [1,2,1,3]}

d2 = {"Profesión": ["Médico", "Ingeniero"],
       "id_profesion": [1,2]
      }

df1 = pd.DataFrame(d1, index={1, 2,3,4})
df2 = pd.DataFrame(d2, index={1, 2})

result = pd.merge(df1,df2,how='inner')
result
```
