```python
#se instala el módulo
!pip -q install kagglehub

#se cargan librerías
import kagglehub
import pandas as pd

#se descarga el dataset de kaggle
ruta_dataset = kagglehub.dataset_download("mohankrishnathalla/medical-insurance-cost-prediction")
print(ruta_dataset)

#se crea un dataframe con el archivo csv
df = pd.read_csv(ruta_dataset + "/medical_insurance.csv")

#se observa el dataframe
df.head()
```

```python
# 1- Genere una muestra aleatoria simple sin reemplazo de 1000 casos. 

muestra_1 = df.sample(n=1000, replace=False, random_state=42)
print('Cantidad de casos en la muestra 1:', muestra_1.shape[0])
muestra_1.head()
```

```python
# 2- Genere una muestra estratificada asumiendo que el estrato es el género, de 1000 casos.

muestra_2_partes = []
proporcion_total = len(df)
resto_por_asignar = 1000

for i, (sexo, grupo) in enumerate(df.groupby('sex')):
    if i < df['sex'].nunique() - 1:
        n_sexo = round(len(grupo) * 1000 / proporcion_total)
        resto_por_asignar = resto_por_asignar - n_sexo
    else:
        n_sexo = resto_por_asignar

    muestra_2_partes.append(grupo.sample(n=n_sexo, replace=False, random_state=42))

muestra_2 = pd.concat(muestra_2_partes)

if muestra_2.shape[0] > 1000:
    muestra_2 = muestra_2.sample(n=1000, random_state=42)

print('Cantidad de casos en la muestra 2:', muestra_2.shape[0])
muestra_2['sex'].value_counts()
```

```python
# 3- Genere una muestra por conglomerado tomando como conglomerado a la región, donde cada 
# conglomerado (5 en total) tenga 100 casos.

muestra_3_partes = []

for region, grupo in df.groupby('region'):
    muestra_3_partes.append(grupo.sample(n=100, replace=False, random_state=42))

muestra_3 = pd.concat(muestra_3_partes)

print('Cantidad de casos en la muestra 3:', muestra_3.shape[0])
muestra_3['region'].value_counts()
```

```python
# 4- Genere una nueva muestra aleatoria simple sin reemplazo de 1000 casos. Para ello primero, quite 
# del dataframe original lo casos que ya seleccionó en el punto 1.

resto_df = df.drop(muestra_1.index)
muestra_4 = resto_df.sample(n=1000, replace=False, random_state=42)

print('Cantidad de casos en la muestra 4:', muestra_4.shape[0])
muestra_4.head()
```

```python
# 5- Concatene las muestra generadas en 1 y 4

muestra_final = pd.concat([muestra_1, muestra_4], ignore_index=True)
print('Cantidad de casos en la muestra final:', muestra_final.shape[0])
muestra_final.head()
```
