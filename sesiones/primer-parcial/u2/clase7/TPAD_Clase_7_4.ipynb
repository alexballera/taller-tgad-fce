### **Ejemplo 1**

Obtenemos un pdf para extraer una tabla de datos.

Este lo puede imprtar directamente al colab o descargarlo desde la web.



Como ejemplo, utilizaremos una descarga directa desde la web: https://www.bcra.gob.ar/PublicacionesEstadisticas/informe-inclusion-financiera-012025.asp

Entrando al link, a la derecha encuentra "descargar versión PDF"-> clcik derecho "copiar enlace"

Luego, abra el pdf haciendo directamente click para ver que tabla quiere extraer. Para el ejemplo, elegimos la que se encuentra en la página 5 "Tabla de indicadores | Inclusión financiera en Argentina"

```python
#se descarga un pdf desde la web

!wget --no-check-certificate https://www.bcra.gob.ar/Pdfs/PublicacionesEstadisticas/Informe-inclusion-financiera-octubre-25.pdf
```

```python
#se crea un dataframe con la tabla de datos extraída del pdf
tables = camelot.read_pdf('/content/Informe-inclusion-financiera-octubre-25.pdf', pages="5", flavor="stream", suppress_stdout=True)
```

```python
#se observa la tabla extraída
df1=tables[0].df
df1
```

Resulta necesario formatear los datos para estructurarlos y poder convertirlos en DataFrame, siendo el objeto operable con Python

```python
#se borran las filas 0 (primera) y 2 (tercera)
df2=df1.drop([0,2], axis=0)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
#borrar  columna 1 (segunda)
df2=df2.drop([1], axis=1)
df2.reset_index(drop=True, inplace=True)
df2
```

```python
#reemplazar nombres de columnas por la primer fila

df2=df2.rename(columns=df2.iloc[0]).drop(df2.index[0])
df2.reset_index(drop=True, inplace=True)
df2
```

```python
df2.dtypes
```

al observar que las columnas 1, 2 y 3 las esta leyendo como string, entonces las convertimos a numéricas

pero observamos que el el separador de decimal es una coma. Entonces debemos reemplazarla por un punto y al mismo tiempo podemos definir el formato

```python
df2.columns
```

```python
#reemplazmaos la como por punto
for j in df2.columns:
  df2[j]=df2[j].str.replace(',','.')
```

```python
df2
```

```python
df2.dtypes
```

```python
#cambiamos a tipo float las columnas correspondientes
for i in df2.columns:
  if i!='Indicador':
    df2[i]=df2[i].astype(float)
```

```python
df2.dtypes
```

Supongamos que queremos extrer del Anexo_Aspectos Metodológicos la tabla que referencia nombres de universidades (descargar el word y observar donde se encuentra)

```python
#se instala el módulo
!pip install python-docx
```

```python
#se carga la librería
import docx

from docx import Document


import pandas as pd
```

```python
#se lee el word
document = Document('/content/Anexo_Aspectos Metodol├│gicos 2023.docx')
```

```python
df1=document.tables[5] #debo prestar atención cual es la posición del objeto (tabla) que quiero extraer
df1
```

```python
data = [[cell.text for cell in row.cells] for row in df1.rows]
df = pd.DataFrame(data)
df
```

```python
#borramos la fila 0
df.drop([0], axis=0, inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
#utilizar fila 0 como encabezados de columnas
df.columns=df.iloc[0]
df=df.drop(df.index[0])
df.reset_index(drop=True, inplace=True)
```
