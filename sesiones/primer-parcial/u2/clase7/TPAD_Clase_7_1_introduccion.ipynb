# Cliente

Dispositivo, software que solicita servicios, recursos o datos a un servidor a través de una red. <br>
El cliente inicia la comunicación y el servidor responde a sus solicitudes.

# Servidor

 Sistema (hardware, software o ambos) que proporciona recursos, servicios o funcionalidades a otros dispositivos o programas

# Protocolo HTTP

### Importancia para el Web Scraping

Es esencial entender cómo funcionan las solicitudes y respuestas HTTP porque:

- **Realizar solicitudes**: Usaremos herramientas como `requests` en Python para enviar solicitudes GET y POST a los servidores y extraer los datos que necesitamos.

- **Manejo de respuestas**: Debemos interpretar correctamente el código de estado y el contenido de la respuesta. Si recibimos un 200, podemos proceder a extraer datos; si es un 404, debemos manejar el error adecuadamente.

- **Encabezados y autenticación**: Algunos sitios requieren que incluimos encabezados específicos o que autenticamos nuestra solicitud (por ejemplo, con tokens de acceso) para obtener los datos. Conocer cómo funcionan estos aspectos es crucial para acceder a información restringida.



HTTP (HyperText Transfer Protocol) es un protocolo de comunicación que permite la transferencia de datos en la web. Se basa en un modelo de cliente-servidor, donde el cliente (normalmente un navegador web) solicita recursos y el servidor los proporciona.

## Estructura de una Solicitud HTTP

Una solicitud HTTP se compone de varias partes:

1. **URL (Uniform Resource Locator)**: Es la dirección del recurso que queremos acceder, junto con los parámetro. Por ejemplo, `https://www.ejemplo.com?page=1`.

2. **Método**: Indica la acción que se desea realizar. Los métodos más comunes son:
   - **GET**: Se utiliza para solicitar datos del servidor. Por ejemplo, cuando abrimos una página web, el navegador envía una solicitud GET para obtener el contenido.
   - **POST**: Se usa para enviar datos al servidor, como al enviar un formulario.
<br/><br/>

3. **Encabezados (Headers)**: Proporcionan información adicional sobre la solicitud, como el tipo de contenido que acepta el cliente (`Accept`), información sobre el navegador (`User-Agent`), entre otros.

4. **Cuerpo (Body)**: Para solicitudes POST, es utilizado para enviar datos al servidor.

<br>



```http
GET /index.html HTTP/1.1
Host: www.google.com
User-Agent: Mozilla/5.0 (X11; Linux x86_64; rv:109.0)
Accept: text/html,application/xhtml+xml,application/xml
```

```http
POST /login HTTP/1.1
Host: www.example.com
User-Agent: Mozilla/5.0 (X11; Linux x86_64; rv:109.0)
Content-Type: application/x-www-form-urlencoded; charset=UTF-8
Content-Length: 28

username=john_doe&password=S3cr3tP@ss
```

## Respuesta HTTP

Cuando el servidor recibe una solicitud, envía una respuesta que también tiene una estructura específica:

1. **Código de estado**: Indica el resultado de la solicitud. Algunos códigos comunes son:

2. **Encabezados de respuesta (header)**: Proporcionan información sobre el tipo de contenido devuelto y otros detalles.

3. **Cuerpo de la respuesta (body)**: Contiene los datos solicitados, como el HTML de una página web, un archivo JSON, etc.



```http
HTTP/1.1 200 OK
Content-Type: text/html
Content-Length: 1256
Set-Cookie: session_id=abc123

<!DOCTYPE html>
<html>
  <head>
    <title>Ejemplo</title>
  </head>
  <body>
    <h1>Bienvenido</h1>
    <!-- Contenido de la página... -->
  </body>
</html>
```

## ¿Qué es HTTPS?

La comunicación entre el cliente web y el servidor web está cifrada. Mediante este sistema, se evita que un tercero no autorizado tenga acceso los datos, por ejemplo, observando el tráfico de la red wifi.

El servidor web se autentica enviando un certificado al navegador justo al comienzo de la transmisión de datos, que garantiza la fiabilidad del dominio. Esta medida ayuda a combatir el fraude por parte de páginas web falsas.

# Códigos de Estado HTTP/HTTPS

## 2xx Éxito
| Código | Descripción       |
|--------|-------------------|
| 200    | OK                |
| 201    | Creado            |
| 202    | Aceptado          |
| 204    | Sin Contenido     |

## 4xx Error del Cliente
| Código | Descripción           |
|--------|-----------------------|
| 400    | Solicitud Incorrecta  |
| 401    | No Autorizado         |
| 403    | Prohibido             |
| 404    | No Encontrado         |

## 5xx Error del Servidor
| Código | Descripción                          |
|--------|--------------------------------------|
| 500    | Error Interno del Servidor           |
| 502    | Puerta de Enlace Incorrecta          |
| 503    | Servicio No Disponible               |
| 504    | Tiempo de Espera de la Puerta de Enlace Agotado |


## 3xx Redirección
| Código | Descripción                          |
|--------|--------------------------------------|
| 301    | Movido Permanentemente               |
| 302    | Encontrado (Anteriormente "Movido Temporalmente")|
| 304    | No Modificado                        |

# CSS

CSS, o Cascading Style Sheets (Hojas de Estilo en Cascada), es un lenguaje de diseño utilizado para describir la presentación y el formato de un documento escrito en HTML o XML. CSS permite a los desarrolladores y diseñadores web aplicar estilos, como colores, fuentes y márgenes, a los elementos de una página web de manera separada del contenido estructural.

```css
h1 {
    color: blue;
    font-size: 24px;
}
```

# Javascript

Lenguaje de programación de alto nivel, interpretado y orientado a objetos, ampliamente utilizado para crear interactividad y dinamismo en páginas web. Desarrollado inicialmente por Netscape en 1995, JavaScript ha evolucionado para convertirse en un estándar del desarrollo web y es una de las tecnologías clave que conforman la tríada fundamental del desarrollo web, junto con HTML y CSS.

### Características Principales

**Interactividad**: JavaScript permite a los desarrolladores crear aplicaciones web interactivas que responden a las acciones del usuario, como clics, desplazamientos y entradas de formularios. Esto mejora significativamente la experiencia del usuario.

**Manipulación del DOM**: JavaScript se utiliza para manipular el Document Object Model (DOM), lo que permite modificar el contenido y la estructura de una página web en tiempo real sin necesidad de recargarla.

**Programación Asíncrona**: Gracias a su naturaleza no bloqueante, JavaScript puede realizar operaciones asíncronas, como solicitudes a servidores, permitiendo a los desarrolladores cargar datos y actualizarlos dinámicamente en la interfaz de usuario.

# Document Object Model - DOM

El Document Object Model (DOM) es una interfaz de programación que proporciona una representación estructurada de documentos HTML y XML. Permite a los lenguajes de programación, como JavaScript, interactuar con la estructura del documento, manipular su contenido y aplicar estilos dinámicamente. El DOM trata un documento como una jerarquía de nodos, donde cada elemento, atributo y texto es un objeto que puede ser accedido y modificado.
