**Detectar valores negativos en columnas numéricas**

Si la variable admite valores negatvos desde su concepto, entonces no será un problema. En caso contrario, si (por ejemplo INGRESO)

```python
for i in DATA_1.columns:
  if DATA_1.dtypes[i] == 'int64' or DATA_1.dtypes[i] == 'float64':
    if (DATA_1[i]<0).any():
      print(f"La columna '{i}' contiene valores negativos")
    else:
      print(f"La columna '{i}' no contiene valores negativos")
```

**Detección de valores atípicos**

Esto es detectar los valores atípicos en base al criterio de BoxPlot: todos aquellos valores que superan los bigotes

```python
for i in DATA_1.columns:
  if DATA_1.dtypes[i] == 'int64' or DATA_1.dtypes[i] == 'float64':
    lim_sup=np.percentile(DATA_1[i],75) + 1.5*(np.percentile(DATA_1[i],75)-np.percentile(DATA_1[i],25))
    lim_inf=np.percentile(DATA_1[i],25) - 1.5*(np.percentile(DATA_1[i],75)-np.percentile(DATA_1[i],25))
    if (DATA_1[i] > lim_sup).any() or (DATA_1[i] < lim_inf).any():
      extreme_values = DATA_1[(DATA_1[i] > lim_sup) | (DATA_1[i] < lim_inf)][i]
      if len(extreme_values.unique())==1:
        print(f"La columna '{i}' contiene '{len(extreme_values.unique())}' valor extremo.")
        print(f"El valor extremo es '{extreme_values.unique()}' ")
      else:
        print(f"La columna '{i}' contiene '{len(extreme_values.unique())}' valores extremos.")
        print(f"Los primeros 10 valores extremos son '{extreme_values.unique()[:10]}' ") #se muestran solo 10
    else:
      print(f"La columna '{i}' no contiene valores extremos.")
```
