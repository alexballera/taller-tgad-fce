```python
#se instala el módulo
!pip -q install kagglehub

#se cargan librerías
import kagglehub
import pandas as pd

#se descarga el dataset de kaggle
ruta_dataset = kagglehub.dataset_download("mohankrishnathalla/medical-insurance-cost-prediction")
print(ruta_dataset)

#se crea un dataframe con el archivo csv
df = pd.read_csv(ruta_dataset + "/medical_insurance.csv")

#se observa el dataframe
df.head()
```

```python
#Rsuelva los siguientes puntos:
# 1- Dimensiones del DataFrame. 

filas, columnas = df.shape
print("Cantidad de filas:", filas)
print("Cantidad de columnas:", columnas)
print("Dimensión (filas, columnas):", df.shape)
```

```python
# 2- Cuente la cantidad de columnas según tipo de dato utilizando FOR e IF

cant_object = 0
cant_int = 0
cant_float = 0
cant_bool = 0
cant_otro = 0

for col in df.columns:
    if df[col].dtype == 'object':
        cant_object = cant_object + 1
    elif df[col].dtype == 'int64':
        cant_int = cant_int + 1
    elif df[col].dtype == 'float64':
        cant_float = cant_float + 1
    elif df[col].dtype == 'bool':
        cant_bool = cant_bool + 1
    else:
        cant_otro = cant_otro + 1

print('Columnas tipo object:', cant_object)
print('Columnas tipo int64:', cant_int)
print('Columnas tipo float64:', cant_float)
print('Columnas tipo bool:', cant_bool)
print('Columnas de otro tipo:', cant_otro)
```

```python
# 3- ¿Cuántas personas hay según zona?

if 'residential_area' in df.columns:
    col_zona = 'residential_area'
elif 'zone' in df.columns:
    col_zona = 'zone'
elif 'region' in df.columns:
    col_zona = 'region'
else:
    col_zona = None
    print('No se encontró una columna de zona en el DataFrame')

if col_zona is not None:
    personas_por_zona = df[col_zona].value_counts()
    print('Columna utilizada:', col_zona)
    print(personas_por_zona)
```

```python
# 4- ¿Cuántas personas tienen más de 50 años?

if 'age' in df.columns:
    col_edad = 'age'
elif 'Age' in df.columns:
    col_edad = 'Age'
else:
    col_edad = None
    print('No se encontró la columna de edad en el DataFrame')

if col_edad is not None:
    mayores_50 = df[df[col_edad] > 50].shape[0]
    print('Cantidad de personas con más de 50 años:', mayores_50)
```

```python
# 5- ¿Cuántas personas tienen más de 50 años y son de la zona rural?

if 'age' in df.columns:
    col_edad = 'age'
elif 'Age' in df.columns:
    col_edad = 'Age'
else:
    col_edad = None
    print('No se encontró la columna de edad en el DataFrame')

if 'residential_area' in df.columns:
    col_zona = 'residential_area'
elif 'zone' in df.columns:
    col_zona = 'zone'
elif 'region' in df.columns:
    col_zona = 'region'
else:
    col_zona = None
    print('No se encontró columna de zona en el DataFrame')

if col_edad is not None and col_zona is not None:
    filtro = (df[col_edad] > 50) & (df[col_zona].astype(str).str.lower() == 'rural')
    cant_mayores_50_rural = df[filtro].shape[0]
    print('Cantidad de personas mayores de 50 años en zona rural:', cant_mayores_50_rural)
```

```python
# 6- ¿Cuál es la mínima y máxima edad que tiene las personas?

if 'age' in df.columns:
    col_edad = 'age'
elif 'Age' in df.columns:
    col_edad = 'Age'
else:
    col_edad = None
    print('No se encontró la columna de edad en el DataFrame')

if col_edad is not None:
    edad_minima = df[col_edad].min()
    edad_maxima = df[col_edad].max()
    print('Edad mínima:', edad_minima)
    print('Edad máxima:', edad_maxima)
```

```python
# 7- Cree una nueva variable utilizando FOR e IF en base a las siguientes reglas:
    #-Si una persona toma alcohol diariamente o semanalmente y fuma -> ‘Riesgo Alto’; 
    #-Si una persona toma alcohol ocasionalmente y fuma -> ‘Riesgo Medio’;
    #-Si una persona toma alcohol diariamente o semanalmente y no fuma -> ‘Riesgo Medio’;
    #-Si una persona toma alcohol ocasionalmente y no fuma -> ‘Riesgo Bajo’;
    #-Si una persona toma alcohol diariamente o semanalmente y es ex fumador -> ‘Riesgo Medio’;
    #-Si una persona toma alcohol ocasionalmente y es ex fumador -> ‘Riesgo Bajo’; 
    #- Si hay alguna otra opción, deberá definir la regla en base a los criterios anteriores

#buscar columnas posibles de manera flexible
col_alcohol = None
col_fuma = None

for c in df.columns:
    c_norm = str(c).lower().replace(' ', '_').replace('-', '_')

    if ('alcohol' in c_norm or 'drink' in c_norm or 'bebida' in c_norm) and col_alcohol is None:
        col_alcohol = c

    if ('smok' in c_norm or 'fum' in c_norm or 'tobac' in c_norm) and col_fuma is None:
        col_fuma = c

if col_alcohol is None or col_fuma is None:
    print('No se encontraron columnas necesarias para crear la variable riesgo')
    print('Columnas disponibles:')
    print(list(df.columns))
else:
    riesgo = []

    for i in range(len(df)):
        alcohol = str(df.loc[i, col_alcohol]).lower().strip()
        fuma = str(df.loc[i, col_fuma]).lower().strip()

        #normalizamos categorías de alcohol
        if 'daily' in alcohol or 'diario' in alcohol or 'weekly' in alcohol or 'semanal' in alcohol:
            alcohol_cat = 'frecuente'
        elif 'occasional' in alcohol or 'ocasional' in alcohol:
            alcohol_cat = 'ocasional'
        else:
            alcohol_cat = 'otro'

        #normalizamos categorías de tabaquismo
        if ('smoker' in fuma or 'fuma' in fuma) and 'ex' not in fuma and 'non' not in fuma and 'no' not in fuma:
            fuma_cat = 'fuma'
        elif 'ex' in fuma or 'former' in fuma:
            fuma_cat = 'ex'
        elif 'non' in fuma or 'no' in fuma or 'nonsmoker' in fuma or 'non-smoker' in fuma:
            fuma_cat = 'no_fuma'
        else:
            fuma_cat = 'otro'

        #reglas de riesgo con FOR e IF
        if alcohol_cat == 'frecuente' and fuma_cat == 'fuma':
            riesgo.append('Riesgo Alto')
        elif alcohol_cat == 'ocasional' and fuma_cat == 'fuma':
            riesgo.append('Riesgo Medio')
        elif alcohol_cat == 'frecuente' and fuma_cat == 'no_fuma':
            riesgo.append('Riesgo Medio')
        elif alcohol_cat == 'ocasional' and fuma_cat == 'no_fuma':
            riesgo.append('Riesgo Bajo')
        elif alcohol_cat == 'frecuente' and fuma_cat == 'ex':
            riesgo.append('Riesgo Medio')
        elif alcohol_cat == 'ocasional' and fuma_cat == 'ex':
            riesgo.append('Riesgo Bajo')
        else:
            #regla adicional para casos no contemplados explícitamente
            if fuma_cat == 'fuma':
                riesgo.append('Riesgo Medio')
            elif fuma_cat == 'ex':
                riesgo.append('Riesgo Bajo')
            else:
                riesgo.append('Riesgo Bajo')

    df['riesgo'] = riesgo
    print('Columna de alcohol usada:', col_alcohol)
    print('Columna de tabaquismo usada:', col_fuma)
    print('Variable creada: riesgo')
    print(df['riesgo'].value_counts())
    df[['riesgo', col_alcohol, col_fuma]].head()
```

```python
# 8- ¿Hay valores faltantes? Si los hay, en que variable/s? utilice FOR e IF

hay_faltantes = False

for col in df.columns:
    cant_nulos = df[col].isna().sum()
    if cant_nulos > 0:
        hay_faltantes = True
        print('Variable con faltantes:', col, '| Cantidad:', cant_nulos)

if hay_faltantes == False:
    print('No hay valores faltantes en el DataFrame')
```

```python
# 9- ¿ Hay valores atípicos? Si los hay, en que variable/s? utilice FOR e IF

columnas_numericas = []

for col in df.columns:
    if df[col].dtype == 'int64' or df[col].dtype == 'float64':
        columnas_numericas.append(col)

for col in columnas_numericas:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lim_inf = q1 - 1.5 * iqr
    lim_sup = q3 + 1.5 * iqr

    cantidad_atipicos = 0
    for valor in df[col]:
        if valor < lim_inf or valor > lim_sup:
            cantidad_atipicos = cantidad_atipicos + 1

    if cantidad_atipicos > 0:
        print('Variable con atípicos:', col, '| Cantidad:', cantidad_atipicos)
```

```python
# 10- Todos los valores de las variables categóricas, ¿son único? utilice FOR e IF

columnas_categoricas = []

for col in df.columns:
    if df[col].dtype == 'object':
        columnas_categoricas.append(col)

for col in columnas_categoricas:
    valores_unicos = []
    for valor in df[col]:
        if valor not in valores_unicos:
            valores_unicos.append(valor)

    if len(valores_unicos) == df[col].shape[0]:
        print('Todos los valores son únicos en la variable:', col)
    else:
        print('La variable', col, 'no tiene todos los valores únicos')
```
