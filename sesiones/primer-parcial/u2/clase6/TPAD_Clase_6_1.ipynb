### **E. conexión a API de Kaggle para descarga de datos**

1. Ingresar a la web de Kaggle: https://www.kaggle.com/

2. Ir a DATASETS y buscar y seleccionar un conjunto de datos

3. Ingresar al dataset seleccionado y en el botón de Download click derecho, copiar el código y ejecutarlo en Colab

Por ejemplo, se elije este: https://www.kaggle.com/datasets/parthpatel2130/realistic-loan-approval-dataset-us-and-canada

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("parthpatel2130/realistic-loan-approval-dataset-us-and-canada")

print("Path to dataset files:", path)
```

Tomar el paht completo que muestra "Path to dataset files:" y agregar el nonbre del archivo que se quiere leer y .csv

El nombre se encuentra en la misma web arriba de donde se visualizan las variables

```python
import pandas as pd
```

```python
import os
import glob

csv_files = glob.glob(os.path.join(path, "*.csv"))

if not csv_files:
    raise FileNotFoundError(f"No se encontró ningún archivo .csv en: {path}")

print("CSV encontrado:", csv_files[0])
DATA = pd.read_csv(csv_files[0])
DATA
```
