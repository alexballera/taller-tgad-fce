**Medidas de tendencia central:**

**MEDIA ARIMÉTICA**: Es el promedio simple de los valores de la variable aletoria Es muy sensible a valores atípicos siendo aquellos que se alejan del comportamiento “normal" de los datos. Su cálculo es:

\begin{align}
         \bar{X} = {\sum_{i=1}^n x_i\over n}
\end{align}

**MEDIANA**: Es el valor central (o mitad que es equivalente al 50%) cuando los datos se encuentran ordenados de manera creciente. Su cálculo permite obtener la posición donde se encuentra el valor dentro de la distribución de los datos:

\begin{align}
         X_{me}
\end{align}

Recordamos que en este caso se aplicabla la siguiente regla (siendo lo que realiza Python para el calculo, también en el caso de los percentiles):

Para obtener su valor se debe primero calcular:
Distancia=k*(n-1) donde k es el porcentaje que en este caso es 0,5

Si:
- Distancia=número entero => es la posición. Buscaré en el ´índice de mis datos dicha posición y me fijare el valor que toma la variable siendo este el valor de la mediana
- Distancia=número decimal => debo interpolar . El resultado de esta operación será el valor de la mediana

**MODA**: Es el valor más frecuente, es decir, el que más veces se repite dentro de la distribución. Si dos o mas valores son los de mayor frecuencia de aparición entonces estamos frente a una distribución multimodal:

\begin{align}
        x_{mo}
\end{align}

**Medidas de Tendencia no central**

- Percentiles: permiten saber el porcentaje de observaciones que se encuentran por debajo de cierto valor y por encima al dividir la distribución de los datos en 100 partes iguales. Su cálculo, que brinda una posición dentro de la distribución de los valores de la variable aletoria, es:

\begin{align}
         {P_k} = {n * k\over 100}
\end{align}

Donde $D_i$ refiere a decil siendo aquel que divide al distribución de los datos en 10 partes iguales, y $Q_i$ refiere a cuartil siendo aquel que divide a la distribución de los datos en 4 partes iguales.

**Medidas de dispersión:**

- RANGO: permite saber la amplitud del dominio de la distribución de los datos. Su cálculo es:

\begin{align}
         {Rango} = { X_{max} - X_{min}}
\end{align}

- RANGO INTERCUARTÍLICO: amplitud del intervalo donde se acumula el 50% de la distribución de los datos. Su cálculo es:

\begin{align}
         {RIC} = { Q_{3} - Q_{1}}
\end{align}

- VARIANZA MUESTRAL: indica cuanto se alejan los datos del valor medio de la distribución en términos cuadráticos. Su cálculo es:

\begin{align}
        S^2= {\sum_{i=1}^n (x_i - \bar{X})^2\over n-1}
\end{align}

- DESVÍO MUESTRAL: indica cuanto se alejan los datos del valor medio de la distribución en la misma unidad de media que la variable. Su cálculo es:

\begin{align}
        S= {\sqrt {S^2}}
\end{align}

- COEFICIENTE DE VARIACIÓN: mide la relación entre el desvío muestral y la media muestral. Su cálculo es:

\begin{align}
        CV= {S \over \bar{X}}
\end{align}

**Medidas de forma**

**COEFICIENTE DE ASIMETRÍA**: permite saber cuan dispersos se encuentran los datos respecto respecto de la media. Su cálculo es:

\begin{align}
         {A_s} = {{\sum_{i=1}^n {\left( (x_i - \bar{X}) \over {S} \right)}^3} \over n}
\end{align}

Interpretación:

- Si ${A_s} < 0 $ entonces la distribución de los datos es asimétrica negativa (concentración de datos a derecha y cola a izquierda). Entonces se cumple que ${\bar{X} < X_{me}}$

- Si ${A_s} = 0 $ entonces la distribución de los datos es simétrica. Entonces se cumple que ${\bar{X} = X_{me}}$

- Si ${A_s} > 0 $ entonces la distribución de los datos es asimétrica positiva (concentración de datos a izquierda y cola a derecha). Entonces se cumple que ${\bar{X} > X_{me}}$

**COEFICIENTE DE CURTOSIS**: permite saber como es la forma de la concentración de los datos. Su cálculo es:

\begin{align}
         {K} = {{\sum_{i=1}^n {\left( (x_i - \bar{X}) \over {S} \right)}^4} \over n}
\end{align}

Interpretación:

- Si ${K} < 3 $ entonces la distribución es platicúrtica (la forma de la distribución es más aplanada).

- Si ${K} = 3 $ entonces la distribución es mesocúrtica (la forma de la distribución no es ni muy aplanada ni muy puntiaguda).

- Si ${K} > 3 $ entonces la distribución es leptocúrtica (la forma de la distribución presenta un pico en el centro).

Las medidas de forma suelen apreciarse a través de gráficos. Particularmente, en estadística hay dos principales: histograma y boxplot. Estos ayudan a visualizar de una manera resumida varias de las medidas anteriores.

###**Covarianza y Correlación lineal**

**Interpretación de la Covarianza:**

- La Cov(X,Y) > 0 si la relación entre X e Y es fuertemente positiva.
 Esto es, a valores grandes de X suceden conjuntamente con
valores grandes de Y (y viceversa).

- La Cov(X,Y) < 0 si la relación entre X e Y es fuertemente negativa.
Esto es, a valores grandes de X suceden conjuntamente
con valores pequeños de Y (y viceversa).

- La Cov(X,Y)=0 no existe relación entre X e Y

**Interpretación de la correlación lineal**

- R=1 relación lineal perfectamente positiva. Esto sucederá cuando X e Y estén fuertemente  relacionadas de manera positiva.

- R= -1 relación lineal perfectamente negativa. Esto sucederá cuando X e Y estén fuertemente relacionadas de manera negativa.

- R=0  indica que existe una ausencia completa de relación lineal entre X e Y.

- 1<R<0 indica que la relación es lineal negativa:
	- Si -1<R≤ -0,8 entonces relación lineal fuertemente negativa
	- Si -0,8<R ≤ -0,5 entonces relación lineal moderadamente negativa
	- Si -0,5<R< 0 entonces relación lineal débilmente negativa

- 0<R<1 indica que la relación es lineal positiva:
	- Si 0<R< 0,5 entonces relación lineal débilmente positiva
	- Si 0<R< 0,8 entonces relación lineal moderadamente positiva
	- Si 0,8<R< 1 entonces relación lineal fuertemente positiva

###**Intervalos de Confianza**

Al trabajar con muestras, resulta importante conocer cuando cerca o lejos nos encontramos respecto del verdadero valor de la media o varianza de cada variable, principalmente

```python
#se crean funciones para el cálculo de los intervalos
def IC_MEDIA_VARDESCON(media_muestral, desv_muestral, N, alfa):
  import scipy.stats as ss
  ESTADÍSTICO= ss.t.ppf(1-(alfa/2), N-1)
  Li=media_muestral - ESTADÍSTICO*(desv_muestral/(N**0.5))
  Ls=media_muestral + ESTADÍSTICO*(desv_muestral/(N**0.5))
  if Li>Ls:
    intervalo='ERROR'
  else:
    INTERVALO=(Li, Ls)
    from IPython.display import display, Math
    intervalo_media=display(Math('IC['+str(round(INTERVALO[0],4))+'\leq \mu \leq'+ str(round(INTERVALO[1],4))+']='+str(1-alfa)))
  return intervalo_media


def IC_VARIANZA(media_muestral, desv_muestral, N, alfa):
   import scipy.stats as ss
   ESTADÍSTICO_1= ss.chi2.ppf(1-(alfa/2), N-1)
   ESTADÍSTICO_2= ss.chi2.ppf((alfa/2), N-1)
   Li=(N-1)*(desv_muestral**2) / ESTADÍSTICO_1
   Ls=(N-1)*(desv_muestral**2) / ESTADÍSTICO_2
   if Li <0 or Li>Ls:
     intervalo='ERROR'
   else:
     INTERVALO=(Li, Ls)
     from IPython.display import display, Math
     intervalo_var=display(Math('IC['+str(round(INTERVALO[0],4))+'\leq \sigma^2 \leq'+ str(round(INTERVALO[1],4))+']='+str(1-alfa)))
   return intervalo_var
```

```python
for i in DATA:
  if DATA[i].dtype=='int64' or DATA[i].dtype=='float64':
    print(f"El intervalo de confianza para la media de la variable {i} es:")
    IC_MEDIA_VARDESCON(DATA[i].mean(), DATA[i].std(),len(DATA[i]),0.05)
    print(f"El intervalo de confianza para la varianza de la variable {i} es:")
    IC_VARIANZA(DATA[i].mean(), DATA[i].std(),len(DATA[i]),0.05)
```

###**Test de hipótesis**

Al trabajar con muestras, también nos puede interesar validar algún suposición sobre la verdadera media o varianza de cada variable, principalmente

```python
def TH_MEDIA_VARDESCON(media_muestral, mu, S, n, alfa, simbolo_H1):
  import scipy.stats as ss
  import numpy as np
  tobs=round((media_muestral-mu) / (S/np.sqrt(n)),4)
  if simbolo_H1=='>':
    talfa  = round(ss.t.ppf(1-alfa, n-1),4)
    if tobs > talfa:
      respuesta='Como se cumple C.D ya que ' + 'tobs='+str(tobs)+' es mayor que '+ 't_1-alfa;n-1='+str(talfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'tobs='+str(tobs)+' es menor que '+ 't_1-alfa;n-1='+str(talfa)+' entonces no se rechaza H0'

    p_valor=round(1-ss.t.cdf(tobs, n-1),8)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  elif simbolo_H1=='<':
    talfa  = round(ss.t.ppf(alfa, n-1),4)
    if tobs < talfa:
      respuesta='Como se cumple C.D ya que ' + 'tobs='+str(tobs)+' es menor que '+ 'talfa;n-1='+str(talfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'tobs='+str(tobs)+' es mayor que '+ 'talfa;n-1='+str(talfa)+' entonces no se rechaza H0'

    p_valor=round(ss.t.cdf(tobs, n-1),8)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  else:
    talfa  = round(ss.t.ppf(1-alfa/2, n-1),4)
    if np.abs(tobs) > talfa:
      respuesta='Como se cumple C.D ya que ' + '|tobs|='+'|'+str(tobs)+'|'+' es mayor que '+ 't_1-alfa/2;n-1='+str(talfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + '|tobs|='+'|'+str(tobs)+'|'+' es menor que '+ 't_1-alfa/2;n-1='+str(talfa)+' entonces no se rechaza H0'

    p_valor=2*round(1-ss.t.cdf(abs(tobs), n-1),8)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  return respuesta, respuesta1

def TH_MEDIA_VARIANZA(var_muestral, var_pob, n, alfa, simbolo_H1):
  import scipy.stats as ss
  import numpy as np
  chiobs=round((n-1)*var_muestral / var_pob,4)
  if simbolo_H1=='>':
    chialfa  = round(ss.chi2.ppf(1-alfa, n-1),4)
    if chiobs > chialfa:
      respuesta='Como se cumple C.D ya que ' + 'chiobs='+str(chiobs)+' es mayor que '+ 'Chi_1-alfa;n-1='+str(chialfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'chiobs='+str(chiobs)+' es menor que '+ 'Chi_1-alfa;n-1='+str(chialfa)+' entonces no se rechaza H0'

    p_valor=round(1-ss.chi2.cdf(chiobs, n-1),8)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  elif simbolo_H1=='<':
    chialfa  = round(ss.chi2.ppf(alfa, n-1),4)
    if chiobs < chialfa:
      respuesta='Como se cumple C.D ya que ' + 'chiobs='+str(chiobs)+' es menor que '+ 'chialfa;n-1='+str(chialfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + 'chiobs='+str(chiobs)+' es mayor que '+ 'chialfa;n-1='+str(chialfa)+' entonces no se rechaza H0'

    p_valor=round(ss.chi2.cdf(chiobs, n-1),8)
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  else:
    chialfa  = round(ss.chi2.ppf(1-alfa/2, n-1),4)
    if np.abs(chiobs) > chialfa:
      respuesta='Como se cumple C.D ya que ' + '|chiobs|='+'|'+str(chiobs)+'|'+' es mayor que '+ 'Chi_1-alfa/2;n-1='+str(chialfa)+' entonces se rechaza H0'
    else:
      respuesta='Como no se cumple C.D ya que ' + '|chiobs|='+'|'+str(chiobs)+'|'+' es menor que '+ 'Chi_1-alfa/2;n-1='+str(chialfa)+' entonces no se rechaza H0'

    p_valor=2*np.min([round(1-ss.chi2.cdf(chiobs, n-1),8), round(ss.chi2.cdf(chiobs, n-1),8)])
    if p_valor < alfa:
      respuesta1='Como se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es menor que '+ 'alfa='+ str(alfa)+' entonces se rechaza H0'
    else:
      respuesta1='Como no se cumple C.D ya que ' + 'P-valor='+str(p_valor)+  ' es mayor que '+ 'alfa='+ str(alfa)+' entonces no se rechaza H0'

  return respuesta, respuesta1
```
